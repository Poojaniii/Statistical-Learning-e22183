Q 1.8

We compute the Minimum Mean Square Error (MMSE), which represents the optimal error (irreducible uncertainty) when using the estimate \(\mathbb{E}[X\mid Y]\).

As derived previously,

$$
\mathrm{MSE}_{\min}
=
\mathbb{E}\!\left[(X-\mathbb{E}[X\mid Y])^2\right]
=
\mathbb{E}[\mathrm{Var}(X\mid Y)].
$$

First, we find the conditional variance for each risk class \(y\in\{0,1\}\).

Since \(X\) is a Bernoulli indicator variable given \(Y=y\), the variance is

$$
p(1-p).
$$

For \(Y=0\),

$$
\mathrm{Var}(X\mid Y=0)
=
\frac{1}{6}\left(1-\frac{1}{6}\right)
=
\frac{1}{6}\cdot\frac{5}{6}
=
\frac{5}{36}.
$$

For \(Y=1\),

$$
\mathrm{Var}(X\mid Y=1)
=
\frac{1}{2}\left(1-\frac{1}{2}\right)
=
\frac{1}{2}\cdot\frac{1}{2}
=
\frac{1}{4}.
$$

Now, we take the expectation over the distribution of \(Y\):

$$
\begin{aligned}
\mathbb{E}[\mathrm{Var}(X\mid Y)]
&=
\mathrm{Var}(X\mid Y=0)\,P(Y=0)
+
\mathrm{Var}(X\mid Y=1)\,P(Y=1)
\\[6pt]
&=
\left(\frac{5}{36}\cdot 0.60\right)
+
\left(\frac{1}{4}\cdot 0.40\right)
\\[6pt]
&=
\left(\frac{5}{36}\cdot\frac{3}{5}\right)
+
\left(\frac{1}{4}\cdot\frac{2}{5}\right)
\\[6pt]
&=
\frac{1}{12}
+
\frac{1}{10}
=
\frac{5+6}{60}
=
\frac{11}{60}
\approx 0.1833.
\end{aligned}
$$

Thus,

$$
\boxed{
\mathrm{MSE}_{\min}
=
\frac{11}{60}
\approx 0.1833
}
$$

This value, \(0.1833\), is the lowest possible average squared error the bank can achieve.

Any other estimate based on risk class would result in an error strictly greater than this.

Q 1.9

If the bank ignores the risk class \(Y\) and uses the global mean

$$
\mathbb{E}[X]=0.30

as its estimate, the resulting error is the \textbf{total variance of \(X\)}.

\textbf{Calculating the Total MSE}

When the estimate is a constant

$$
c=\mathbb{E}[X],

the Mean Square Error is simply the variance of \(X\):

$$
\mathrm{MSE}_{\text{global}}
=
\mathbb{E}\!\left[(X-\mathbb{E}[X])^2\right]
=
\mathrm{Var}(X).
$$

Since \(X\) is a Bernoulli random variable with

$$
P(X=1)=0.30,
$$

we have

$$
\mathrm{Var}(X)
=
p(1-p)
=
0.30(0.70)
=
0.21.
$$

\textbf{Comparing the Errors}

\begin{itemize}
    \item \textbf{Optimal Error (with \(Y\)):}
    $$
    \mathrm{MSE}_{\min}
    =
    \mathbb{E}[\mathrm{Var}(X\mid Y)]
    =
    \frac{11}{60}
    \approx 0.1833
    $$

    \item \textbf{Global Error (without \(Y\)):}
    $$
    \mathrm{Var}(X)
    =
    \frac{12.6}{60}
    =
    0.21
    $$
\end{itemize}

By using the information in \(Y\), the bank reduces its prediction error from

$$
0.21
\quad \text{to} \quad
0.1833.
$$

\textbf{Relationship to the Law of Total Variance}

The \textbf{Law of Total Variance} decomposes the total uncertainty into two parts:

$$
\mathrm{Var}(X)
=
\mathbb{E}[\mathrm{Var}(X\mid Y)]
+
\mathrm{Var}(\mathbb{E}[X\mid Y]).
$$

In this example:

\begin{itemize}
    \item
    $$
    \mathbb{E}[\mathrm{Var}(X\mid Y)]
    =
    0.1833
    $$
    is the \textbf{unexplained variance}. It represents the uncertainty that remains even after using the best predictor.

    \item
    $$
    \mathrm{Var}(\mathbb{E}[X\mid Y])
    $$
    is the \textbf{explained variance}. It measures how much prediction error is reduced by using \(Y\).
\end{itemize}

Now compute it.

Since

$$
\mathbb{E}[X\mid Y]
=
\begin{cases}
\dfrac{1}{6}, & \text{with probability } 0.6, \\[6pt]
\dfrac{1}{2}, & \text{with probability } 0.4,
\end{cases}
$$

we have

$$
\mathrm{Var}(\mathbb{E}[X\mid Y])
=
\mathbb{E}\!\left[(\mathbb{E}[X\mid Y])^2\right]
-
(\mathbb{E}[X])^2.
$$

Therefore,

$$
\begin{aligned}
\mathrm{Var}(\mathbb{E}[X\mid Y])
&=
\left[
\left(\frac{1}{36}\cdot 0.6\right)
+
\left(\frac{1}{4}\cdot 0.4\right)
\right]
-
(0.3)^2
\\[6pt]
&=
[0.0167 + 0.1] - 0.09
\\[6pt]
&=
0.0267.
\end{aligned}
$$

\textbf{Final Breakdown}

$$
0.21
=
0.1833
+
0.0267
$$

or equivalently,

$$
\underbrace{0.21}_{\text{Total Error}}
=
\underbrace{0.1833}_{\text{Remaining Error}}
+
\underbrace{0.0267}_{\text{Error Reduction}}.
$$

\textbf{Summary}

The Law of Total Variance shows that the optimal prediction error is always smaller than the global prediction error.

The difference between the two is exactly

$$
\mathrm{Var}(\mathbb{E}[X\mid Y]),
$$

which measures how much predictive power the variable \(Y\) provides.

To calculate these information-theoretic quantities, we first summarize the joint distribution \(P(X,Y)\) and the conditional probabilities.

\section*{1. Joint Distribution \(P(X,Y)\)}

Based on the mappings \(X(\omega)\) and \(Y(\omega)\), we obtain

\begin{align*}
P(X=0,\;Y=0) &= P(\{\omega_1\}) = 0.50, \\
P(X=1,\;Y=0) &= P(\{\omega_2\}) = 0.10, \\
P(X=0,\;Y=1) &= P(\{\omega_3\}) = 0.20, \\
P(X=1,\;Y=1) &= P(\{\omega_4\}) = 0.20.
\end{align*}

The marginal distributions are

\textbf{For \(Y\):}

$$
P(Y=0)=0.60,
\qquad
P(Y=1)=0.40.
$$

\textbf{For \(X\):}

$$
P(X=0)=0.70,
\qquad
P(X=1)=0.30.
$$

\section*{2. Conditional Entropy \(H(X\mid Y)\)}

Conditional entropy is the weighted average of the entropy of \(X\) within each class of \(Y\).

\subsection*{Step A: Entropy of \(X\) given \(Y=0\)}

$$
P(X=1\mid Y=0)=\frac{1}{6},
\qquad
P(X=0\mid Y=0)=\frac{5}{6}.
$$

Therefore,

$$
H(X\mid Y=0)
=
-\left[
\frac{1}{6}\log_2\frac{1}{6}
+
\frac{5}{6}\log_2\frac{5}{6}
\right]
\approx 0.650 \text{ bits}.
$$

\subsection*{Step B: Entropy of \(X\) given \(Y=1\)}

$$
P(X=1\mid Y=1)=\frac{1}{2},
\qquad
P(X=0\mid Y=1)=\frac{1}{2}.
$$

Hence,

$$
H(X\mid Y=1)
=
-\left[
\frac{1}{2}\log_2\frac{1}{2}
+
\frac{1}{2}\log_2\frac{1}{2}
\right]
=
1.00 \text{ bit}.
$$

\subsection*{Step C: Final Conditional Entropy}

$$
H(X\mid Y)
=
P(Y=0)H(X\mid Y=0)
+
P(Y=1)H(X\mid Y=1).
$$

Thus,

$$
\begin{aligned}
H(X\mid Y)
&=
(0.60\times 0.650)
+
(0.40\times 1.00)
\\[6pt]
&=
0.390 + 0.40
\\[6pt]
&=
0.790 \text{ bits}.
\end{aligned}
$$

\section*{3. Mutual Information \(I(X;Y)\)}

Mutual information measures the reduction in uncertainty:

$$
I(X;Y)=H(X)-H(X\mid Y).
$$

\subsection*{Step A: Calculate \(H(X)\)}

$$
H(X)
=
-\left(
0.70\log_2 0.70
+
0.30\log_2 0.30
\right)
\approx 0.881 \text{ bits}.
$$

\subsection*{Step B: Subtract \(H(X\mid Y)\)}

$$
I(X;Y)
=
0.881 - 0.790
=
0.091 \text{ bits}.
$$

\section*{4. Interpretation}

\begin{itemize}
    \item
    $$
    H(X)=0.881
    $$
    is the total uncertainty about fraud before observing the risk class.

    \item
    $$
    H(X\mid Y)=0.790
    $$
    is the uncertainty remaining after the bank knows whether a customer is low-risk or high-risk.

    \item
    $$
    I(X;Y)=0.091
    $$
    is the information gain provided by \(Y\).
\end{itemize}

By knowing the customer's risk class, the bank reduces its uncertainty about fraud by approximately

$$
\frac{0.091}{0.881}\times 100\%
\approx 10.3\%.
$$

\textbf{Practical Interpretation}

Since

$$
I(X;Y)>0,
$$

the variable \(Y\) provides useful predictive information about the fraud indicator \(X\).

Thus, the customer's risk class is a meaningful feature for the bank's fraud detection model.

**QUESTION 2**

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.



To derive the marginal density of $\mathbf{X}$ (or $\mathbf{X}$) from the joint density of $(\mathbf{X}, \mathbf{Y})$, we must integrate the joint density over all possible values of $\mathbf{Y}$ (or $\mathbf{X}$ to find the marginal of  $\mathbf{Y}$).

The challenge lies in the quadratic form within the exponential. To solve this rigorously, we use the **block matrix inversion** (Schur complement) and the **completing the square** technique.


From the definition the joint density $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y}) \propto \exp(-\frac{1}{2} Q)$, where the quadratic form $Q$ is:
$$Q = \begin{bmatrix} \mathbf{x} - \boldsymbol{\mu}_x \\ \mathbf{y} - \boldsymbol{\mu}_y \end{bmatrix}^T \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}^{-1} \begin{bmatrix} \mathbf{x} - \boldsymbol{\mu}_x \\ \mathbf{y} - \boldsymbol{\mu}_y \end{bmatrix}$$

Using the **Schur Complement** of $\boldsymbol{\Sigma}_{XX}$, the inverse of the block covariance matrix $\boldsymbol{\Sigma}^{-1}$ can be written as:
$$\boldsymbol{\Sigma}^{-1} = \begin{bmatrix} \mathbf{I} & -\boldsymbol{\Sigma}_{XX}^{-1}\boldsymbol{\Sigma}_{XY} \\ \mathbf{0} & \mathbf{I} \end{bmatrix}^T \begin{bmatrix} \boldsymbol{\Sigma}_{XX}^{-1} & \mathbf{0} \\ \mathbf{0} & \mathbf{S}_{YY}^{-1} \end{bmatrix} \begin{bmatrix} \mathbf{I} & -\boldsymbol{\Sigma}_{XX}^{-1}\boldsymbol{\Sigma}_{XY} \\ \mathbf{0} & \mathbf{I} \end{bmatrix}$$
where $\mathbf{S}_{YY} = \boldsymbol{\Sigma}_{YY} - \boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{XX}^{-1}\boldsymbol{\Sigma}_{XY}$.



Letting $\widetilde{\mathbf{x}} = \mathbf{x} - \boldsymbol{\mu}_x$ and $\widetilde{\mathbf{y}} = \mathbf{y} - \boldsymbol{\mu}_y$, we can decompose the exponent $Q$ into two independent parts:
$$Q = \widetilde{\mathbf{x}}^T \boldsymbol{\Sigma}_{XX}^{-1} \widetilde{\mathbf{x}} + (\widetilde{\mathbf{y}} - \mathbf{A}\widetilde{\mathbf{x}})^T \mathbf{S}_{YY}^{-1} (\widetilde{\mathbf{y}} - \mathbf{A}\widetilde{\mathbf{x}})$$
where $\mathbf{A} = \boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{XX}^{-1}$.

The joint density now factors as:
$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y}) = C \cdot \exp\left( -\frac{1}{2} \widetilde{\mathbf{x}}^T \boldsymbol{\Sigma}_{XX}^{-1} \widetilde{\mathbf{x}} \right) \cdot \exp\left( -\frac{1}{2} (\widetilde{\mathbf{y}} - \mathbf{m})^T \mathbf{S}_{YY}^{-1} (\widetilde{\mathbf{y}} - \mathbf{m}) \right)$$
where $\mathbf{m} = \mathbf{A}\widetilde{\mathbf{x}}$ and $C$ is the normalization constant.

---

The marginal density $f_{\mathbf{X}}(\mathbf{x})$ is defined as:
$$f_{\mathbf{X}}(\mathbf{x}) = \int_{\mathbb{R}^m} f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y}) d\mathbf{y}$$

Because the first exponential term depends only on $\mathbf{x}$, it can be pulled out of the integral:
$$f_{\mathbf{X}}(\mathbf{x}) = C \cdot e^{-\frac{1}{2} \widetilde{\mathbf{x}}^T \boldsymbol{\Sigma}_{XX}^{-1} \widetilde{\mathbf{x}}} \int_{\mathbb{R}^m} \exp\left( -\frac{1}{2} (\widetilde{\mathbf{y}} - \mathbf{m})^T \mathbf{S}_{YY}^{-1} (\widetilde{\mathbf{y}} - \mathbf{m}) \right) d\mathbf{y}$$

The integral is simply the kernel of a multivariate normal distribution for $\mathbf{Y}$ (centered at $\mathbf{m}$ with covariance $\mathbf{S}_{YY}$). By the property of probability densities, this integral evaluates to the reciprocal of the normalization constant for $\mathbf{Y}$:
$$\int_{\mathbb{R}^m} (\dots) d\mathbf{y} = \sqrt{(2\pi)^m |\mathbf{S}_{YY}|}$$

---

Using the determinant identity for block matrices, $|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{XX}| \cdot |\mathbf{S}_{YY}|$. When we combine the constants:
$$f_{\mathbf{X}}(\mathbf{x}) = \frac{\sqrt{(2\pi)^m |\mathbf{S}_{YY}|}}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}_{XX}||\mathbf{S}_{YY}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}_x)^T \boldsymbol{\Sigma}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}_x) \right)$$

Simplifying the constants yields the **Marginal Density of X**:
$$f_{\mathbf{X}}(\mathbf{x}) = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}_x)^T \boldsymbol{\Sigma}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}_x) \right)$$

This proves that $\mathbf{X} \sim N(\boldsymbol{\mu}_x, \boldsymbol{\Sigma}_{XX})$.

The derivation for $f_{\mathbf{Y}}(\mathbf{y})$ is mathematically symmetric to the derivation for $\mathbf{X}$. The goal is to integrate out $\mathbf{x}$ to isolate the distribution of $\mathbf{Y}$.


We use the **Schur complement of $\boldsymbol{\Sigma}_{YY}$** (instead of $\boldsymbol{\Sigma}_{XX}$) to factor the quadratic form $Q$. The inverse covariance matrix $\boldsymbol{\Sigma}^{-1}$ is decomposed such that the terms involving $\mathbf{x}$ are "completed" into a square:

$$Q = \widetilde{\mathbf{y}}^T \boldsymbol{\Sigma}_{YY}^{-1} \widetilde{\mathbf{y}} + (\widetilde{\mathbf{x}} - \mathbf{B}\widetilde{\mathbf{y}})^T \mathbf{S}_{XX}^{-1} (\widetilde{\mathbf{x}} - \mathbf{B}\widetilde{\mathbf{y}})$$

Where:
* $\widetilde{\mathbf{y}} = \mathbf{y} - \boldsymbol{\mu}_y$
* $\mathbf{B} = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}$ (the regression coefficients of $\mathbf{X}$ on $\mathbf{Y}$)
* $\mathbf{S}_{XX} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$ (the Schur complement of $\boldsymbol{\Sigma}_{YY}$)

---

The marginal $f_{\mathbf{Y}}(\mathbf{y})$ is calculated by integrating the joint density over $d\mathbf{x}$:

$$f_{\mathbf{Y}}(\mathbf{y}) = \int_{\mathbb{R}^n} f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y}) d\mathbf{x}$$

The term $\exp\left(-\frac{1}{2} \widetilde{\mathbf{y}}^T \boldsymbol{\Sigma}_{YY}^{-1} \widetilde{\mathbf{y}}\right)$ is constant with respect to $\mathbf{x}$ and moves outside the integral. The remaining integral over $\mathbb{R}^n$ covers the kernel of a Gaussian in $\mathbf{X}$, which evaluates to $\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}$.


Using the determinant identity $|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{YY}| \cdot |\mathbf{S}_{XX}|$, the constants cancel out just as before, leaving:

$$f_{\mathbf{Y}}(\mathbf{y}) = \frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}} \exp\left( -\frac{1}{2} (\mathbf{y} - \boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_y) \right)$$



To derive the conditional density $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$, we use the definition of conditional probability for densities:

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}$$

Since we have already decomposed the joint density and the marginal density of $\mathbf{Y}$ in the previous steps, this becomes a task of algebraic subtraction in the exponent.

---

From our previous derivation, we factored the joint quadratic form $Q$ using the Schur complement of $\boldsymbol{\Sigma}_{YY}$:
$$Q_{joint} = (\mathbf{y} - \boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_y) + (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*)$$

Where:
* $\mathbf{S}_{XX} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

The marginal density $f_{\mathbf{Y}}(\mathbf{y})$ has the exponent:
$$Q_{marginal} = (\mathbf{y} - \boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_y)$$

When we divide the densities, we subtract the exponents ($e^{A} / e^{B} = e^{A-B}$):
$$Q_{conditional} = Q_{joint} - Q_{marginal} = (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*)$$

---

Now we look at the coefficients.
* **Joint constant:** $\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}}$
* **Marginal constant:** $\frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}}$

Dividing the joint constant by the marginal constant gives:
$$\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \cdot \sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|} = \frac{1}{\sqrt{(2\pi)^n \frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|}}}$$

Using the determinant identity $|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{YY}| \cdot |\mathbf{S}_{XX}|$, the term $\frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|}$ simplifies exactly to $|\mathbf{S}_{XX}|$.

---

Combining the simplified constant and the remaining quadratic form, we get:

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right)$$

This is the density of a multivariate normal distribution:
$$(\mathbf{X} \mid \mathbf{Y} = \mathbf{y}) \sim \mathscr{N}(\boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y), \quad \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}).$$

---

**Summary of the Parameters**
* **Conditional Mean:** $\boldsymbol{\mu}_{X|Y} = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$
    * *Insight:* The mean is adjusted by how much $\mathbf{y}$ deviates from its expected value, scaled by the "regression" coefficient $\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}$.
* **Conditional Covariance:** $\boldsymbol{\Sigma}_{X|Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
    * *Insight:* Notice this does **not** depend on the specific value of $\mathbf{y}$. The variance is always reduced (or stays the same) compared to the marginal $\boldsymbol{\Sigma}_{XX}$.

Does the fact that the conditional variance is independent of the observed value $y$ spark any ideas regarding its application in linear regression or Kalman filters?

Let $g(y) = \mathbb{E}[X \mid Y=y]$. By the Disintegration Theorem, this is computed as the integral against the conditional PDF:
$$g(y) = \int_{\mathbb{R}^n} \mathbf{x} f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) \, d\mathbf{x}.$$
This integral is simply the first moment (mean) of a Gaussian random variable
$$(\mathbf{X} \mid \mathbf{Y} = \mathbf{y}) \sim \mathscr{N}(\boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y), \quad \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}).$$


Thus:
$$g(y) = \mathbf{\mu}(\mathbf{y}) = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y).$$

Therefore:
$$\mathbb{E}[X \mid Y=y] = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y).$$

and hence
$$\mathbb{E}[X \mid Y] = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{Y} - \boldsymbol{\mu}_y)$$



Since $$(\mathbf{X} \mid \mathbf{Y} = \mathbf{y}) \sim N(\boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y), \quad \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX})$$

$$\mathrm{Var}(\mathbf{X} \mid \mathbf{Y}=y) = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}.$$
SInce this is independent of $y$ we have that
$$\mathrm{Var}(\mathbf{X} \mid \mathbf{Y}) =\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$$
and hence
$$\mathbb{E}\left(\mathrm{Var}(\mathbf{X} \mid \mathbf{Y}) \right)=\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}.$$

Thus we finally have that the estimation error is:
\begin{align*}
\mathbb{E}[(\mathbf{X} - \mathbb{E}[\mathbf{X} \mid \mathbf{Y}])^2] &= \mathbb{E}[\text{Var}(\mathbf{X} \mid \mathbf{Y})]=\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
\end{align*}

We have seen that for any two random variables $\mathbf{X}$ and $\mathbf{Y}$ with finite variance, the total variance of $X$ can be split into two distinct parts:

$$\mathrm{Var}(\mathbf{X}) = \mathbb E[\mathrm{Var}(\mathbf{X}|\mathbf{Y})] + \mathrm{Var}(\mathbb E[\mathbf{X}|\mathbf{Y}]).$$

Thus we see that
$$\mathrm{Var}(\mathbb E[\mathbf{X}|\mathbf{Y}])=\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}.$$

>**Explained Variance:** $\text{Var}(\mathbb{E}[X|Y]) = \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$.
This represents the part of $X$'s variance that is "accounted for" or explained by the linear relationship with $Y$.  
**Unexplained Variance (Error):** $\text{Var}(X|Y) = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$.
This is the remaining uncertainty in $X$ after $Y$ is known, often referred to as the estimation error or residual variance.

Consider the multivariate normal random variable $\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$ with mean $\boldsymbol{\mu} \in \mathbb{R}^{n+m}$ and covariance matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$. The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given as:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right)$$

The differential entropy $h(\mathbf{Z})$ is defined as the expected value of the negative log-density:


$$h(\mathbf{Z}) = -\mathbb{E}[\ln f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})]$$

**Step A: Take the natural logarithm of the density**
Using the joint density formula:


$$\ln f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = -\frac{1}{2} \ln\left((2\pi)^{n+m} |\boldsymbol{\Sigma}|\right) - \frac{1}{2}(\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu})$$

**Step B: Apply the expectation**


$$h(\mathbf{Z}) = \mathbb{E}\left[ \frac{1}{2} \ln\left((2\pi)^{n+m} |\boldsymbol{\Sigma}|\right) + \frac{1}{2}(\mathbf{Z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{Z} - \boldsymbol{\mu}) \right]$$

Since the first term is a constant with respect to the expectation, we have:


$$h(\mathbf{Z}) = \frac{1}{2} \ln\left((2\pi)^{n+m} |\boldsymbol{\Sigma}|\right) + \frac{1}{2} \mathbb{E}\left[ (\mathbf{Z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{Z} - \boldsymbol{\mu}) \right]$$

**Step C: Evaluate the quadratic form expectation**
For any multivariate normal distribution with covariance $\boldsymbol{\Sigma}$ , the expectation of the quadratic form $(\mathbf{Z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{Z} - \boldsymbol{\mu})$ is equal to the dimension of the vector $\mathbf{Z}$, which is $n+m$:


$$\mathbb{E}\left[ (\mathbf{Z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{Z} - \boldsymbol{\mu}) \right] = \text{Tr}(\boldsymbol{\Sigma}^{-1} \boldsymbol{\Sigma}) = \text{Tr}(\mathbf{I}_{n+m}) = n+m$$

**Step D: Final Simplification**
Substitute the result back into the entropy equation:


$$h(\mathbf{Z}) = \frac{1}{2} \ln\left((2\pi)^{n+m} |\boldsymbol{\Sigma}|\right) + \frac{n+m}{2}$$

By writing $\frac{n+m}{2}$ as $\frac{1}{2} \ln(e^{n+m})$, we combine the terms into the final entropy expression:


$$h(\mathbf{Z}) = \frac{1}{2} \ln\left((2\pi e)^{n+m} |\boldsymbol{\Sigma}|\right)$$

The conditional entropy $h(\mathbf{X}\mid\mathbf{Y})$ represents the average amount of uncertainty remaining in $\mathbf{X}$ after $\mathbf{Y}$ is known. For a multivariate Gaussian, this is determined by the varaince of the contional distribution
$$(\mathbf{X} \mid \mathbf{Y} = \mathbf{y}) \sim \mathscr{N}(\boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y), \quad \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX})$$

The conditional covariance $\Sigma_{X|Y}$ covariance is thus given by $$\boldsymbol{\Sigma}_{\mathbf{X}|\mathbf{Y}} = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}.$$  
Since $\boldsymbol{\Sigma}_{\mathbf{X}|\mathbf{Y}}$ is independent of the specific value $y$, the uncertainty is constant across all observations of $Y$.  The differential entropy for this conditional distribution is thus :
$$h(\mathbf{X}\mid\mathbf{Y}) = \frac{1}{2} \ln((2\pi e)^n |\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}).$$

From the law of total variance we have seen that the explained varaince is:
$\text{Var}(\mathbb{E}[\mathbf{X}\mid\mathbf{Y}]) = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$. Thus the differential entropy of $\mathbb{E}(\mathbf{X}\mid\mathbf{Y})$is:
$$h(\mathbb{E}(\mathbf{X}\mid\mathbf{Y})) = \frac{1}{2} \ln((2\pi e)^n |\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}|).$$

Mutual information is given by:

$$I(\mathbf{X};\mathbf{Y}) = h(\mathbf{X}) - h(\mathbf{X} \mid \mathbf{Y}).$$

Substitute the entropy formulas for the marginal and conditional distributions derived in the document:

$$I(\mathbf{X};\mathbf{Y}) = \frac{1}{2} \ln\left((2\pi e)^n |\boldsymbol{\Sigma}_{XX}|\right) - \frac{1}{2} \ln\left((2\pi e)^n |\boldsymbol{\Sigma}_{\mathbf{X}|\mathbf{Y}}|\right) \text{.}$$

Using the property of logarithms ($\ln A - \ln B = \ln \frac{A}{B}$), the constants $(2\pi e)^n$ cancel out:

$$I(\mathbf{X};\mathbf{Y}) = \frac{1}{2} \ln \left( \frac{|\boldsymbol{\Sigma}_{XX}|}{|\boldsymbol{\Sigma}_{\mathbf{X}|\mathbf{Y}}|} \right) \text{.}$$

Substituting the expression for the conditional covariance $\boldsymbol{\Sigma}_{\mathbf{X}|\mathbf{Y}} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$:

$$I(\mathbf{X};\mathbf{Y}) = \frac{1}{2} \ln \left( \frac{|\boldsymbol{\Sigma}_{XX}|}{|\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}|} \right).$$

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# ----------------------------
# 1. PARAMETERS
# ----------------------------
mu = np.array([0, 0])

sigma = np.array([
    [1.0, 0.7],
    [0.7, 1.0]
])

# ----------------------------
# 2. DATA GENERATION
# ----------------------------
n_samples = 5000
data = np.random.multivariate_normal(mu, sigma, n_samples)

x_samples = data[:, 0]
y_samples = data[:, 1]

# ----------------------------
# 3. CONDITIONAL SETUP
# ----------------------------
y_fixed = 1.2

mu_cond = mu[0] + (sigma[0, 1] / sigma[1, 1]) * (y_fixed - mu[1])
var_cond = sigma[0, 0] - (sigma[0, 1] ** 2 / sigma[1, 1])

condition_mask = np.abs(y_samples - y_fixed) < 0.1
x_conditional_samples = x_samples[condition_mask]

# ----------------------------
# 4. FIGURE LAYOUT (CLEAN DARK ACADEMIC STYLE)
# ----------------------------
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        "Joint Distribution",
        "Marginal of Y",
        "Marginal of X",
        f"Conditional X | Y = {y_fixed}"
    ),
    vertical_spacing=0.12,
    horizontal_spacing=0.1
)

# ----------------------------
# 5. JOINT DISTRIBUTION (SMOOTH + CLEAN)
# ----------------------------
fig.add_trace(
    go.Histogram2dContour(
        x=x_samples,
        y=y_samples,
        colorscale="IceFire",
        contours=dict(showlabels=True),
        name="Joint",
        showscale=False
    ),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(
        x=x_samples,
        y=y_samples,
        mode="markers",
        marker=dict(size=2, opacity=0.15, color="white"),
        name="Samples"
    ),
    row=1, col=1
)

fig.add_hline(
    y=y_fixed,
    line_dash="dash",
    line_color="red",
    row=1, col=1
)

# ----------------------------
# 6. MARGINAL Y (CLEAN BLUE)
# ----------------------------
fig.add_trace(
    go.Histogram(
        y=y_samples,
        marker=dict(color="#4C78A8"),
        opacity=0.75,
        nbinsy=40,
        name="Marginal Y"
    ),
    row=1, col=2
)

# ----------------------------
# 7. MARGINAL X (CLEAN GREEN)
# ----------------------------
fig.add_trace(
    go.Histogram(
        x=x_samples,
        marker=dict(color="#59A14F"),
        opacity=0.75,
        nbinsx=40,
        name="Marginal X"
    ),
    row=2, col=1
)

# ----------------------------
# 8. CONDITIONAL DISTRIBUTION (HIGHLIGHT RED)
# ----------------------------
fig.add_trace(
    go.Histogram(
        x=x_conditional_samples,
        marker=dict(color="#E15759"),
        opacity=0.85,
        nbinsx=30,
        name="Conditional"
    ),
    row=2, col=2
)

# ----------------------------
# 9. FINAL LAYOUT POLISH
# ----------------------------
fig.update_layout(
    height=850,
    width=1000,
    title=dict(
        text="Multivariate Normal: Joint, Marginal & Conditional Structure",
        x=0.5,
        font=dict(size=20)
    ),
    paper_bgcolor="#0f1117",
    plot_bgcolor="#0f1117",
    font=dict(color="white"),
    margin=dict(l=40, r=40, t=80, b=40),
    showlegend=False
)

# Axis styling (VERY important for aesthetics)
fig.update_xaxes(showgrid=False, zeroline=False)
fig.update_yaxes(showgrid=False, zeroline=False)

fig.show()
```

**QUESTION 3**

**1. Joint Distribution of $(X, Y)$:**

Since $X$ and $Q$ are independent normal random variables, their sum $Y = X + Q$ is also normally distributed.
* **Mean:** $E[Y] = E[X] + E[Q] = 0 + 0 = 0$.
* **Variance:** Since $X \perp Q$, $\text{Var}(Y) = \text{Var}(X) + \text{Var}(Q) = 1 + 4 = 5$.
* **Covariance:** $\text{Cov}(X, Y) = E[XY] - E[X]E[Y] = E[X(X+Q)] = E[X^2] + E[XQ]$.
    * Since $E[X^2] = \text{Var}(X) = 1$ and $E[XQ] = E[X]E[Q] = 0$:
    * **$\text{Cov}(X, Y) = 1$**.

The joint distribution is:
$$\begin{pmatrix} X \\ Y \end{pmatrix} \sim \mathscr{N} \left( \begin{pmatrix} 0 \\ 0 \end{pmatrix}, \begin{pmatrix} 1 & 1 \\ 1 & 5 \end{pmatrix} \right)$$

---

**2. Information Geometry:**

$X$ is not $\sigma(Y)$-measurable because there is no deterministic function $g$ such that $X = g(Y)$. For any observed $Y=y$, there are infinitely many pairs $(x, q)$ such that $x+q=y$.
**Practically:** A high power reading ($Y$) is ambiguous; it could be a "Thermal Fault" (high $X$) or simply a "Hard Material" (high $Q$). The sensor $Y$ cannot distinguish between these two physical states.

---

**3. Conditional Expectation (The Best Estimate):**

For bivariate normal variables, the best mean-square estimate is:
$$E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y) = 0 + \frac{1}{5}(y - 0)$$
$$\mathbf{E[X \mid Y] = 0.2 Y}$$
**Interpretation:** Because the noise ($Q$) is 4 times more volatile than the signal ($X$), we only "trust" 20% of the power fluctuation as being caused by temperature.

---

**4. Verification of the Residual:**

Let $\widehat{X} = 0.2 Y$. The residual is $Z = X - 0.2 Y$.
* **Mean:** $E[Z] = E[X] - 0.2 E[Y] = 0 - 0 = 0$.
* **Orthogonality:** $\text{Cov}(Z, Y) = E[ZY] = E[(X - 0.2Y)Y] = E[XY] - 0.2 E[Y^2]$.
    * $E[XY] = 1$ and $E[Y^2] = 5$.
    * $\text{Cov}(Z, Y) = 1 - 0.2(5) = 1 - 1 = 0$.
The error is uncorrelated with our observation, meaning we have extracted all possible information from $Y$.

---

**5. Differential Entropy:**

Recall $h(\mathscr{N}(\mu, \sigma^2)) = \frac{1}{2}\ln(2\pi e \sigma^2)$.
* **Marginal $h(X)$:** With $\sigma_X^2 = 1$:
$$h(X) = \frac{1}{2}\ln(2\pi e) \approx 1.4189 \text{ nats}$$
* **Conditional $h(X|Y)$:** The conditional variance is
$$\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)} = 1 - \frac{1}{5} = 0.8.$$
$$h(X|Y) = \frac{1}{2}\ln(2\pi e \cdot 0.8) \approx 1.3073 \text{ nats}$$

---

**6. Mutual Information:**

$$I(X; Y) = h(X) - h(X|Y) = \frac{1}{2}\ln(1) - \frac{1}{2}\ln(0.8) = \frac{1}{2}\ln\left(\frac{1}{0.8}\right) = \frac{1}{2}\ln(1.25) \approx \mathbf{0.1116 \text{ nats}}$$
**Percentage Uncertainty Removed:** This is often measured by the reduction in variance. We went from a variance of $1.0$ to $0.8$ (a **20% reduction**). If the noise $Q$ increased to $\sigma_Q^2 = 100$, $I(X;Y)$ would approach 0, as the temperature signal would be "drowned out."

---

**7. Diagnostic Interpretation:**

The manager's suggestion to use $X = Y$ is a **Naive Estimator**.
* If we use $X = Y$, our error variance would be $E[(X-Y)^2] = E[Q^2] = 4$.
* By using $E[X|Y] = 0.2Y$, our error variance is only $0.8$.

**Shrinkage:** The factor $0.2$ is called a shrinkage estimator. It tells us that when a sensor is noisy, the "best" guess is to be conservative and pull the reading back toward the historical average (zero). If the power spikes, we don't immediately assume the tool is melting; we assume most of that spike is just the material being tough.

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from scipy.stats import multivariate_normal

# --- 1. PARAMETERS ---
mu_x, mu_y = 0, 0
var_x, var_q = 1, 4
var_y = var_x + var_q
cov_xy = var_x
k = cov_xy / var_y  # 0.2

# --- 2. DATA ---
np.random.seed(42)
n_samples = 1000

x_samples = np.random.normal(0, np.sqrt(var_x), n_samples)
q_samples = np.random.normal(0, np.sqrt(var_q), n_samples)
y_samples = x_samples + q_samples

x_hat = k * y_samples
residual = x_samples - x_hat

# --- GRID ---
x_grid = np.linspace(-4, 4, 200)
y_grid = np.linspace(-8, 8, 200)
X_m, Y_m = np.meshgrid(x_grid, y_grid)

pos = np.dstack((X_m, Y_m))
Z = multivariate_normal(
    [mu_x, mu_y],
    [[var_x, cov_xy], [cov_xy, var_y]]
).pdf(pos)

# --- 3. FIGURE (NEON DARK STYLE) ---
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=(
        "Signal Extraction in Joint Space",
        "Orthogonality of Estimation Error"
    )
)

# =========================
# LEFT: JOINT DISTRIBUTION
# =========================

fig.add_trace(
    go.Contour(
        x=x_grid,
        y=y_grid,
        z=Z,
        colorscale="Electric",
        opacity=0.85,
        showscale=False
    ),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(
        x=x_samples,
        y=y_samples,
        mode="markers",
        marker=dict(
            size=3,
            color="rgba(255,255,255,0.35)"
        ),
        name="Samples"
    ),
    row=1, col=1
)

# Optimal estimator
y_line = np.linspace(-8, 8, 100)

fig.add_trace(
    go.Scatter(
        x=k * y_line,
        y=y_line,
        line=dict(color="#00FFB3", width=4),
        name="E[X|Y]"
    ),
    row=1, col=1
)

# Naive estimator
fig.add_trace(
    go.Scatter(
        x=y_line,
        y=y_line,
        line=dict(color="#FF4D4D", width=2, dash="dash"),
        name="Naive X=Y"
    ),
    row=1, col=1
)

# =========================
# RIGHT: RESIDUAL PLOT
# =========================

fig.add_trace(
    go.Scatter(
        x=y_samples,
        y=residual,
        mode="markers",
        marker=dict(
            size=4,
            color="rgba(180,100,255,0.5)"
        ),
        name="Residual"
    ),
    row=1, col=2
)

fig.add_trace(
    go.Scatter(
        x=y_line,
        y=np.zeros_like(y_line),
        line=dict(color="#FFFFFF", width=2),
        name="Zero Line"
    ),
    row=1, col=2
)

# =========================
# FINAL AESTHETICS
# =========================

fig.update_layout(
    title=dict(
        text="Spindle Motor System: Signal Extraction Under Noise",
        x=0.5,
        font=dict(size=20, color="white")
    ),
    paper_bgcolor="#0B0F1A",
    plot_bgcolor="#0B0F1A",
    font=dict(color="white"),
    height=650,
    width=1100,
    showlegend=False
)

fig.update_xaxes(showgrid=False, zeroline=False)
fig.update_yaxes(showgrid=False, zeroline=False)

fig.show()
```
