

# Q. Analytical Derivation

**Filter model:**

$$x_k^- = A_{k-1} x_{k-1}^+ + G_{k-1} w_{k-1}$$
$$y_k^- = H_k x_k^- + z_k$$
$$x_{k-1}^+ \sim \mathcal{N}(m_{k-1}, P_{k-1}), \quad w_k \sim \mathcal{N}(0, \Sigma_p), \quad z_k \sim \mathcal{N}(0, \Sigma_m)$$

all mutually independent.

---

## 1. Show that x⁻ₖ ~ N(m⁻ₖ, P⁻ₖ)

x⁻ₖ is an **affine function of two independent Gaussian random variables** x⁺_{k-1} and w_{k-1}, so it is itself Gaussian.

**Mean** — taking expectation of the state equation:

$$m_k^- \triangleq \mathbb{E}[x_k^-] = \mathbb{E}[A_{k-1} x_{k-1}^+ + G_{k-1} w_{k-1}]$$
$$= A_{k-1} \mathbb{E}[x_{k-1}^+] + G_{k-1} \mathbb{E}[w_{k-1}]$$
$$= A_{k-1} m_{k-1} + G_{k-1} \cdot 0 = A_{k-1} m_{k-1} \quad \checkmark$$

**Covariance** — applying the variance operator, using independence of x⁺_{k-1} and w_{k-1}:

$$P_k^- \triangleq \text{Var}(x_k^-) = \text{Var}(A_{k-1} x_{k-1}^+ + G_{k-1} w_{k-1})$$
$$= A_{k-1} \text{Var}(x_{k-1}^+) A_{k-1}^T + G_{k-1} \text{Var}(w_{k-1}) G_{k-1}^T$$
$$= A_{k-1} P_{k-1} A_{k-1}^T + G_{k-1} \Sigma_p G_{k-1}^T \quad \checkmark$$

Therefore:
$$x_k^- \sim \mathcal{N}(m_k^-, P_k^-) \quad \text{with} \quad m_k^- = A_{k-1} m_{k-1}, \quad P_k^- = A_{k-1} P_{k-1} A_{k-1}^T + G_{k-1} \Sigma_p G_{k-1}^T$$

---

## 2. Show that y⁻ₖ ~ N(Hₖ m⁻ₖ, Hₖ P⁻ₖ Hᵀₖ + Σₘ)

y⁻ₖ = Hₖ x⁻ₖ + zₖ is an **affine function of independent Gaussians** x⁻ₖ ~ N(m⁻ₖ, P⁻ₖ) and zₖ ~ N(0, Σₘ), so it is Gaussian.

**Mean:**

$$\mathbb{E}[y_k^-] = H_k \mathbb{E}[x_k^-] + \mathbb{E}[z_k] = H_k m_k^- + 0 = H_k m_k^- \quad \checkmark$$

**Variance** — using independence of x⁻ₖ and zₖ:

$$\text{Var}(y_k^-) = H_k \text{Var}(x_k^-) H_k^T + \text{Var}(z_k) = H_k P_k^- H_k^T + \Sigma_m \quad \checkmark$$

Therefore:
$$y_k^- \sim \mathcal{N}(H_k m_k^-,\; H_k P_k^- H_k^T + \Sigma_m)$$

---

## 3. Show the Joint Distribution

Write the joint vector as $[x_k^- ; y_k^-] = M [x_k^- ; z_k]$ where:

$$M = \begin{bmatrix} I & 0 \\ H_k & I \end{bmatrix}$$

Since $[x_k^- ; z_k]$ is jointly Gaussian with mean $[m_k^- ; 0]$ and block-diagonal covariance $\text{diag}(P_k^-, \Sigma_m)$ (by independence), applying the linear map M gives:

**Joint mean:**
$$\mathbb{E}\begin{bmatrix} x_k^- \\ y_k^- \end{bmatrix} = \begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix} \quad \checkmark$$

**Cross-covariance:**
$$\text{Cov}(x_k^-, y_k^-) = \text{Cov}(x_k^-,\; H_k x_k^- + z_k) = \text{Cov}(x_k^-, x_k^-) H_k^T + \text{Cov}(x_k^-, z_k)$$
$$= P_k^- H_k^T + 0 = P_k^- H_k^T \quad \checkmark$$

**Full joint covariance** $\Lambda = M \cdot \text{diag}(P_k^-, \Sigma_m) \cdot M^T$:

$$\Lambda = \begin{bmatrix} P_k^- & P_k^- H_k^T \\ H_k P_k^- & H_k P_k^- H_k^T + \Sigma_m \end{bmatrix}$$

Therefore:
$$\begin{bmatrix} x_k^- \\ y_k^- \end{bmatrix} \sim \mathcal{N}\!\left( \begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix},\; \begin{bmatrix} P_k^- & P_k^- H_k^T \\ H_k P_k^- & H_k P_k^- H_k^T + \Sigma_m \end{bmatrix} \right) \quad \checkmark$$

---

## 4. Show the Posterior x⁺ₖ | yᵒᵇˢₖ ~ N(mₖ, Pₖ)

Apply the **standard Gaussian conditioning formula** to the joint distribution from (3). Partition:

$$\mu_1 = m_k^-, \quad \Sigma_{11} = P_k^-$$
$$\mu_2 = H_k m_k^-, \quad \Sigma_{22} = H_k P_k^- H_k^T + \Sigma_m, \quad \Sigma_{12} = P_k^- H_k^T$$

**Conditional mean:**

$$m_k = \mu_1 + \Sigma_{12} \Sigma_{22}^{-1}(y_k^{\text{obs}} - \mu_2)$$
$$= m_k^- + P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1}(y_k^{\text{obs}} - H_k m_k^-)$$

Defining the **Kalman gain**:
$$K_k \triangleq P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1}$$

we get:
$$m_k = m_k^- + K_k(y_k^{\text{obs}} - H_k m_k^-) \quad \checkmark$$

**Conditional covariance:**

$$P_k = \Sigma_{11} - \Sigma_{12} \Sigma_{22}^{-1} \Sigma_{21}$$
$$= P_k^- - P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1} H_k P_k^-$$
$$= P_k^- - K_k H_k P_k^-$$
$$= (I - K_k H_k) P_k^- \quad \checkmark$$

Therefore $x_k^+ \mid y_k^{\text{obs}} \sim \mathcal{N}(m_k, P_k)$.

---

## 5. Find E[x⁻ₖ | y⁻ₖ = yᵒᵇˢₖ] and Var(x⁻ₖ | y⁻ₖ = yᵒᵇˢₖ)

These follow **directly from (4)**:

$$\mathbb{E}[x_k^- \mid y_k^- = y_k^{\text{obs}}] = m_k = m_k^- + K_k(y_k^{\text{obs}} - H_k m_k^-) \quad \checkmark$$

$$\text{Var}(x_k^- \mid y_k^- = y_k^{\text{obs}}) = P_k = (I - K_k H_k) P_k^- \quad \checkmark$$

The posterior mean is the prior mean **shifted by the Kalman gain times the innovation** $(y_k^{\text{obs}} - H_k m_k^-)$. The posterior covariance is **strictly reduced** from the prior $P_k^-$ by the factor $(I - K_k H_k)$, reflecting the information gained from the observation.

# Q. 1-D Example

**Scalar model:**

$$x_k^- = a\, x_{k-1}^+ + w_{k-1}, \quad w_{k-1} \sim \mathcal{N}(0, q)$$
$$y_k^- = h\, x_k^- + z_k, \quad z_k \sim \mathcal{N}(0, r)$$
$$x_0 \sim \mathcal{N}(m_0, P_0), \quad Y_k = \{y_1^{\text{obs}}, \ldots, y_k^{\text{obs}}\}$$

---

## 1. Show that m⁻ₖ = a m_{k-1} and P⁻ₖ = a² P_{k-1} + q

This is the scalar special case of the general prediction result (Q. Analytical Derivation, Proof 1), with $A_{k-1} = a$, $G_{k-1} = 1$, $\Sigma_p = q$.

**Mean** — taking expectation of the state equation:

$$m_k^- \triangleq \mathbb{E}[x_k^-] = \mathbb{E}[a\, x_{k-1}^+ + w_{k-1}]$$
$$= a\, \mathbb{E}[x_{k-1}^+] + \mathbb{E}[w_{k-1}]$$
$$= a\, m_{k-1} + 0 = a\, m_{k-1} \quad \checkmark$$

**Variance** — applying the variance operator, using independence of $x_{k-1}^+$ and $w_{k-1}$:

$$P_k^- \triangleq \text{Var}(x_k^-) = \text{Var}(a\, x_{k-1}^+ + w_{k-1})$$
$$= a^2\, \text{Var}(x_{k-1}^+) + \text{Var}(w_{k-1})$$
$$= a^2 P_{k-1} + q \quad \checkmark$$

---

## 2. Show that mₖ and Pₖ

Define the **innovation** and **innovation variance**:

$$v_k \triangleq y_k^{\text{obs}} - h\, m_k^-, \qquad S_k \triangleq h^2 P_k^- + r$$

**Kalman gain** (scalar case of $K_k = P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1}$ with $H_k = h$):

$$K_k = \frac{P_k^-\, h}{S_k} = \frac{P_k^-\, h}{h^2 P_k^- + r}$$

**Updated mean:**

$$m_k = m_k^- + K_k v_k = m_k^- + \frac{P_k^- h}{S_k}(y_k^{\text{obs}} - h\, m_k^-) \quad \checkmark$$

**Updated variance:**

$$P_k = (1 - K_k h)\, P_k^- = \left(1 - \frac{P_k^- h^2}{S_k}\right) P_k^- \quad \checkmark$$

Note that since $K_k h = \frac{P_k^- h^2}{S_k} \in (0,1)$, the posterior variance $P_k$ is always strictly less than the prior $P_k^-$.

---

## 3. Show the Predictive Measurement Distribution p(yₖ | Y_{k-1})

Before seeing $y_k^{\text{obs}}$, the predicted state is $x_k^- \sim \mathcal{N}(m_k^-, P_k^-)$, independent of $z_k \sim \mathcal{N}(0, r)$.

Since $y_k^- = h\, x_k^- + z_k$ is an affine function of independent Gaussians:

$$\mathbb{E}[y_k^-] = h\, m_k^-$$
$$\text{Var}(y_k^-) = h^2\, P_k^- + r$$

Therefore:

$$p(y_k \mid Y_{k-1}) = \mathcal{N}(h\, m_k^-,\; h^2 P_k^- + r) \quad \checkmark$$

---

## 4. Show the Posterior-Predictive Distribution p(yₖ | Yₖ)

After incorporating $y_k^{\text{obs}}$, the posterior state is $x_k^+ \sim \mathcal{N}(m_k, P_k)$. A fresh observation at the same step satisfies $y_k = h\, x_k^+ + z_k$ with $z_k \sim \mathcal{N}(0, r)$ independent, giving:

$$\mathbb{E}[y_k] = h\, m_k$$
$$\text{Var}(y_k) = h^2\, P_k + r$$

Therefore:

$$p(y_k \mid Y_k) = \mathcal{N}(h\, m_k,\; h^2 P_k + r) \quad \checkmark$$

Since $P_k < P_k^-$, the posterior-predictive is always **narrower** than the predictive — reflecting the information gained from the observation $y_k^{\text{obs}}$.

---

## 5. Animation — Prior vs Posterior for suitable numerical values

Choose: $a = 0.98,\; q = 0.5,\; h = 1.0,\; r = 2.0,\; m_0 = 0.0,\; P_0 = 5.0$



```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
from scipy.stats import norm

# ── Parameters ────────────────────────────────────────────────
a, q, h, r = 0.98, 0.5, 1.0, 2.0
m0, P0     = 0.0, 5.0
T, seed    = 30, 42

# ── Simulate true state and noisy observations ────────────────
rng = np.random.default_rng(seed)
x     = np.zeros(T)
y_obs = np.zeros(T)
x[0]  = rng.normal(m0, np.sqrt(P0))
for k in range(T - 1):
    x[k+1] = a * x[k] + rng.normal(0, np.sqrt(q))
for k in range(T):
    y_obs[k] = h * x[k] + rng.normal(0, np.sqrt(r))

# ── Run 1-D Kalman Filter ─────────────────────────────────────
m_preds, P_preds, m_upds, P_upds = [], [], [], []
m, P = m0, P0
for k in range(T):
    # Predict  (Proof 1)
    m_pred = a * m
    P_pred = a**2 * P + q
    # Update   (Proof 2)
    S = h**2 * P_pred + r
    K = P_pred * h / S
    v = y_obs[k] - h * m_pred
    m_upd = m_pred + K * v
    P_upd = (1 - K * h) * P_pred
    m_preds.append(m_pred); P_preds.append(P_pred)
    m_upds.append(m_upd);   P_upds.append(P_upd)
    m, P = m_upd, P_upd

m_preds = np.array(m_preds); P_preds = np.array(P_preds)
m_upds  = np.array(m_upds);  P_upds  = np.array(P_upds)

# Innovation variance for measurement distributions (Proofs 3 & 4)
S_pred = h**2 * P_preds + r    # predictive measurement variance
S_post = h**2 * P_upds  + r    # posterior-predictive measurement variance

# ── Animation ─────────────────────────────────────────────────
xgrid = np.linspace(-10, 10, 400)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("1-D Scalar Kalman Filter: Prior vs Posterior", fontsize=13, fontweight='bold')

def update_frame(k):
    for ax in axes:
        ax.cla()

    # Left panel: state distributions
    prior_pdf = norm.pdf(xgrid, m_preds[k], np.sqrt(P_preds[k]))
    post_pdf  = norm.pdf(xgrid, m_upds[k],  np.sqrt(P_upds[k]))
    axes[0].plot(xgrid, prior_pdf, 'b--', lw=2,
                 label=f'Prior  N({m_preds[k]:.2f}, {P_preds[k]:.2f})')
    axes[0].plot(xgrid, post_pdf,  'g-',  lw=2.5,
                 label=f'Posterior N({m_upds[k]:.2f}, {P_upds[k]:.2f})')
    axes[0].axvline(x[k],     color='r',      lw=1.5, linestyle=':',
                    label=f'True x = {x[k]:.2f}')
    axes[0].axvline(y_obs[k], color='orange', lw=1.5, linestyle='--',
                    label=f'Obs y = {y_obs[k]:.2f}')
    axes[0].fill_between(xgrid, prior_pdf, alpha=0.15, color='blue')
    axes[0].fill_between(xgrid, post_pdf,  alpha=0.20, color='green')
    axes[0].set_title(f'State Distribution  (k = {k+1})')
    axes[0].set_xlabel('x');  axes[0].set_ylabel('p(x)')
    axes[0].legend(fontsize=8); axes[0].set_xlim(xgrid[0], xgrid[-1])
    axes[0].set_ylim(0, None)

    # Right panel: measurement distributions (Proofs 3 & 4)
    pred_pdf      = norm.pdf(xgrid, h * m_preds[k], np.sqrt(S_pred[k]))
    post_pred_pdf = norm.pdf(xgrid, h * m_upds[k],  np.sqrt(S_post[k]))
    axes[1].plot(xgrid, pred_pdf,      'b--', lw=2,
                 label=f'Predictive  N({h*m_preds[k]:.2f}, {S_pred[k]:.2f})')
    axes[1].plot(xgrid, post_pred_pdf, 'g-',  lw=2.5,
                 label=f'Post-pred   N({h*m_upds[k]:.2f}, {S_post[k]:.2f})')
    axes[1].axvline(y_obs[k], color='orange', lw=2,
                    label=f'y_obs = {y_obs[k]:.2f}')
    axes[1].fill_between(xgrid, pred_pdf,      alpha=0.15, color='blue')
    axes[1].fill_between(xgrid, post_pred_pdf, alpha=0.20, color='green')
    axes[1].set_title(f'Measurement Distribution  (k = {k+1})')
    axes[1].set_xlabel('y');  axes[1].set_ylabel('p(y)')
    axes[1].legend(fontsize=8); axes[1].set_xlim(xgrid[0], xgrid[-1])
    axes[1].set_ylim(0, None)

    fig.tight_layout()

ani = FuncAnimation(fig, update_frame, frames=T, interval=350, repeat=True)
ani.save("1d_kf_animation.gif", writer=PillowWriter(fps=3))
plt.show()
```

# Q. 2D Position Estimation — Part A

**State vector** at time step $k$:

$$x_k = \begin{bmatrix} p_x(k) \\ p_y(k) \\ v_x(k) \\ v_y(k) \end{bmatrix}$$

where $(p_x(k), p_y(k))$ are position components and $(v_x(k), v_y(k))$ are velocity components.

**Measurement** (position only):

$$y_k = \begin{bmatrix} p_x^{\text{meas}}(k) \\ p_y^{\text{meas}}(k) \end{bmatrix}$$

---

## Derivation of A

We assume **constant-velocity kinematics** — no acceleration, with sampling period $\Delta t$. The equations of motion are:

$$p_x(k) = p_x(k-1) + \Delta t \cdot v_x(k-1)$$
$$p_y(k) = p_y(k-1) + \Delta t \cdot v_y(k-1)$$
$$v_x(k) = v_x(k-1)$$
$$v_y(k) = v_y(k-1)$$

Writing $x_k = A\, x_{k-1}$ in matrix form:

$$\begin{bmatrix} p_x(k) \\ p_y(k) \\ v_x(k) \\ v_y(k) \end{bmatrix} = \begin{bmatrix} 1 & 0 & \Delta t & 0 \\ 0 & 1 & 0 & \Delta t \\ 0 & 0 & 1 & 0 \\ 0 & 0 & 0 & 1 \end{bmatrix} \begin{bmatrix} p_x(k-1) \\ p_y(k-1) \\ v_x(k-1) \\ v_y(k-1) \end{bmatrix}$$

Therefore:

$$\boxed{A = \begin{bmatrix} 1 & 0 & \Delta t & 0 \\ 0 & 1 & 0 & \Delta t \\ 0 & 0 & 1 & 0 \\ 0 & 0 & 0 & 1 \end{bmatrix}} \quad \checkmark$$

---

## Derivation of H

Only the position components are measured, so $y_k = H x_k$ must extract the first two elements of the state:

$$\begin{bmatrix} p_x^{\text{meas}}(k) \\ p_y^{\text{meas}}(k) \end{bmatrix} = \begin{bmatrix} 1 & 0 & 0 & 0 \\ 0 & 1 & 0 & 0 \end{bmatrix} \begin{bmatrix} p_x(k) \\ p_y(k) \\ v_x(k) \\ v_y(k) \end{bmatrix}$$

Therefore:

$$\boxed{H = \begin{bmatrix} 1 & 0 & 0 & 0 \\ 0 & 1 & 0 & 0 \end{bmatrix}} \quad \checkmark$$

---

## Derivation of G

The process noise models **unmodeled random acceleration**:

$$w_{k-1} = \begin{bmatrix} a_x(k-1) \\ a_y(k-1) \end{bmatrix} \sim \mathcal{N}(0, \Sigma_p)$$

Over one sampling interval $\Delta t$, a random acceleration affects both position (second-order) and velocity (first-order) via kinematic integration:

$$\Delta p_x = \frac{1}{2}\Delta t^2\, a_x, \qquad \Delta p_y = \frac{1}{2}\Delta t^2\, a_y$$
$$\Delta v_x = \Delta t\, a_x, \qquad \Delta v_y = \Delta t\, a_y$$

So the full state perturbation is $x_{\text{noise}} = G\, w_{k-1}$:

$$\begin{bmatrix} \Delta p_x \\ \Delta p_y \\ \Delta v_x \\ \Delta v_y \end{bmatrix} = \begin{bmatrix} \frac{1}{2}\Delta t^2 & 0 \\ 0 & \frac{1}{2}\Delta t^2 \\ \Delta t & 0 \\ 0 & \Delta t \end{bmatrix} \begin{bmatrix} a_x \\ a_y \end{bmatrix}$$

Therefore:

$$\boxed{G = \begin{bmatrix} \frac{1}{2}\Delta t^2 & 0 \\ 0 & \frac{1}{2}\Delta t^2 \\ \Delta t & 0 \\ 0 & \Delta t \end{bmatrix}} \quad \checkmark$$

The induced **state-space process noise covariance** is then:

$$Q = G\, \Sigma_p\, G^T$$

With $\Sigma_p = q^2 I_2$ (isotropic acceleration noise), this gives:

$$Q = q^2\, G G^T = q^2 \begin{bmatrix} \frac{1}{4}\Delta t^4 & 0 & \frac{1}{2}\Delta t^3 & 0 \\ 0 & \frac{1}{4}\Delta t^4 & 0 & \frac{1}{2}\Delta t^3 \\ \frac{1}{2}\Delta t^3 & 0 & \Delta t^2 & 0 \\ 0 & \frac{1}{2}\Delta t^3 & 0 & \Delta t^2 \end{bmatrix}$$



**Q. 2D Position Estimation — Part B**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# ─────────────────────────────────────────────────────────────
# SYSTEM MATRICES
# Derived in Part A
# ─────────────────────────────────────────────────────────────

def make_matrices(dt, q, r):
    """
    Build A, H, G, Q, R for the 2D constant-velocity model.

    Parameters
    ----------
    dt : float  - sampling period Δt
    q  : float  - acceleration noise std  (Σp = q²I₂)
    r  : float  - GPS measurement noise std  (R = r²I₂)

    Returns
    -------
    A  : (4,4)  state transition matrix
    H  : (2,4)  measurement matrix
    G  : (4,2)  noise input matrix
    Q  : (4,4)  process noise covariance  Q = G Σp Gᵀ
    R  : (2,2)  measurement noise covariance
    """
    # State transition matrix (Part A)
    A = np.array([[1, 0, dt,  0],
                  [0, 1,  0, dt],
                  [0, 0,  1,  0],
                  [0, 0,  0,  1]], dtype=float)

    # Measurement matrix — position only (Part A)
    H = np.array([[1, 0, 0, 0],
                  [0, 1, 0, 0]], dtype=float)

    # Noise input matrix (Part A)
    G = np.array([[0.5*dt**2,       0],
                  [0,         0.5*dt**2],
                  [dt,              0],
                  [0,              dt]], dtype=float)

    # Process noise covariance  Q = G Σp Gᵀ,  Σp = q²I₂
    Sigma_p = (q**2) * np.eye(2)
    Q = G @ Sigma_p @ G.T

    # Measurement noise covariance  R = r²I₂
    R = (r**2) * np.eye(2)

    return A, H, G, Q, R


# ─────────────────────────────────────────────────────────────
# KALMAN FILTER — PREDICT AND UPDATE
# ─────────────────────────────────────────────────────────────

def kf_predict(m, P, A, Q):
    """
    Prediction step.

    m⁻ₖ = A mₖ₋₁
    P⁻ₖ = A Pₖ₋₁ Aᵀ + Q
    """
    m_pred = A @ m
    P_pred = A @ P @ A.T + Q
    return m_pred, P_pred


def kf_update(m_pred, P_pred, y_obs, H, R):
    """
    Measurement update step.

    Sₖ  = H P⁻ₖ Hᵀ + R          innovation covariance
    Kₖ  = P⁻ₖ Hᵀ Sₖ⁻¹            Kalman gain
    vₖ  = yᵒᵇˢₖ − H m⁻ₖ          innovation
    mₖ  = m⁻ₖ + Kₖ vₖ            updated mean
    Pₖ  = (I − Kₖ H) P⁻ₖ         updated covariance
    """
    S   = H @ P_pred @ H.T + R
    K   = P_pred @ H.T @ np.linalg.inv(S)
    v   = y_obs - H @ m_pred
    m_upd = m_pred + K @ v
    P_upd = (np.eye(len(m_pred)) - K @ H) @ P_pred
    return m_upd, P_upd, K, S


# ─────────────────────────────────────────────────────────────
# SIMULATE GROUND TRUTH + NOISY GPS
# ─────────────────────────────────────────────────────────────

def simulate_trajectory(dt=0.1, r=1.0, T=200, seed=0):
    """
    Simulate a sinusoidal 2D trajectory and add GPS noise.

    Returns
    -------
    X_true : (T,4)  true state  [px, py, vx, vy]
    Y_obs  : (T,2)  noisy GPS position measurements
    t      : (T,)   time vector
    """
    rng = np.random.default_rng(seed)
    t   = np.arange(T) * dt

    # True trajectory
    px = 5.0 * np.sin(0.5 * t)
    py = 0.3 * t
    vx = 5.0 * 0.5 * np.cos(0.5 * t)
    vy = 0.3 * np.ones(T)
    X_true = np.column_stack([px, py, vx, vy])

    # Noisy GPS — position only
    Y_obs = X_true[:, :2] + rng.normal(0, r, size=(T, 2))

    return X_true, Y_obs, t


# ─────────────────────────────────────────────────────────────
# RUN THE FILTER
# ─────────────────────────────────────────────────────────────

def run_kalman_filter(Y_obs, dt=0.1, q=0.01, r=1.0):
    """
    Run the 2D Kalman filter on a sequence of GPS measurements.

    Parameters
    ----------
    Y_obs : (T,2)  noisy GPS position measurements
    dt    : float  sampling period
    q     : float  process noise std
    r     : float  measurement noise std

    Returns
    -------
    ms : (T,4)  filtered state estimates  [px, py, vx, vy]
    Ps : (T,4,4) filtered covariance matrices
    """
    A, H, G, Q, R = make_matrices(dt, q, r)
    T = len(Y_obs)

    # Initialise from first GPS reading, assume zero velocity
    m = np.array([Y_obs[0, 0], Y_obs[0, 1], 0.0, 0.0])
    P = np.diag([r**2, r**2, 1.0, 1.0])

    ms, Ps = [], []
    for k in range(T):
        if k == 0:
            ms.append(m.copy())
            Ps.append(P.copy())
            continue
        m, P = kf_predict(m, P, A, Q)
        m, P = kf_update(m, P, Y_obs[k], H, R)[:2]
        ms.append(m.copy())
        Ps.append(P.copy())

    return np.array(ms), np.array(Ps)


# ─────────────────────────────────────────────────────────────
# PLOT RESULTS
# ─────────────────────────────────────────────────────────────

def plot_uncertainty_ellipse(ax, mean, cov, n_std=1.0, **kwargs):
    """Draw a 1-sigma uncertainty ellipse from a 2x2 covariance."""
    vals, vecs = np.linalg.eigh(cov)
    angle = np.degrees(np.arctan2(vecs[1, 0], vecs[0, 0]))
    w, h  = 2 * n_std * np.sqrt(vals)
    ell   = Ellipse(mean, w, h, angle=angle, **kwargs)
    ax.add_patch(ell)


def plot_results(X_true, Y_obs, ms, Ps, t):

    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    fig.suptitle("2D Constant-Velocity Kalman Filter — GPS Filtering",
                 fontsize=13, fontweight='bold')

    # ── Panel 1: XY Trajectory ───────────────────────────────
    ax = axes[0, 0]
    ax.plot(X_true[:, 0], X_true[:, 1],
            'k-',  lw=2,   label='True trajectory')
    ax.scatter(Y_obs[:, 0], Y_obs[:, 1],
               s=8, c='orange', alpha=0.5, label='Noisy GPS')
    ax.plot(ms[:, 0], ms[:, 1],
            'b-',  lw=2,   label='KF estimate')

    # 1-sigma ellipses every 20 steps
    for i in range(0, len(ms), 20):
        plot_uncertainty_ellipse(
            ax, ms[i, :2], Ps[i][:2, :2],
            edgecolor='blue', facecolor='none', alpha=0.4, lw=1.2)

    ax.set_xlabel('px'); ax.set_ylabel('py')
    ax.set_title('XY Trajectory  (blue ellipses = 1σ uncertainty)')
    ax.legend(fontsize=8); ax.set_aspect('equal'); ax.grid(True, alpha=0.3)

    # ── Panel 2: Position x vs Time ──────────────────────────
    ax = axes[0, 1]
    ax.plot(t, X_true[:, 0], 'k-',     lw=1.5, label='True px')
    ax.scatter(t, Y_obs[:, 0], s=5, c='orange', alpha=0.4, label='GPS px')
    ax.plot(t, ms[:, 0],    'b-',     lw=1.5, label='KF px')
    sig = np.sqrt(np.array([P[0, 0] for P in Ps]))
    ax.fill_between(t, ms[:, 0] - sig, ms[:, 0] + sig,
                    alpha=0.2, color='blue', label='±1σ')
    ax.set_xlabel('time'); ax.set_ylabel('px')
    ax.set_title('Position x  vs  Time')
    ax.legend(fontsize=8); ax.grid(True, alpha=0.3)

    # ── Panel 3: Position y vs Time ──────────────────────────
    ax = axes[1, 0]
    ax.plot(t, X_true[:, 1], 'k-',     lw=1.5, label='True py')
    ax.scatter(t, Y_obs[:, 1], s=5, c='orange', alpha=0.4, label='GPS py')
    ax.plot(t, ms[:, 1],    'b-',     lw=1.5, label='KF py')
    sig = np.sqrt(np.array([P[1, 1] for P in Ps]))
    ax.fill_between(t, ms[:, 1] - sig, ms[:, 1] + sig,
                    alpha=0.2, color='blue', label='±1σ')
    ax.set_xlabel('time'); ax.set_ylabel('py')
    ax.set_title('Position y  vs  Time')
    ax.legend(fontsize=8); ax.grid(True, alpha=0.3)

    # ── Panel 4: Estimation Error ─────────────────────────────
    ax = axes[1, 1]
    err_gps_x = Y_obs[:, 0] - X_true[:, 0]
    err_gps_y = Y_obs[:, 1] - X_true[:, 1]
    err_kf_x  = ms[:, 0]    - X_true[:, 0]
    err_kf_y  = ms[:, 1]    - X_true[:, 1]

    ax.plot(t, np.abs(err_gps_x), color='orange', alpha=0.5, lw=1,   label='|GPS err px|')
    ax.plot(t, np.abs(err_gps_y), color='gold',   alpha=0.5, lw=1,   label='|GPS err py|')
    ax.plot(t, np.abs(err_kf_x),  color='blue',             lw=1.5,  label='|KF err px|')
    ax.plot(t, np.abs(err_kf_y),  color='navy',             lw=1.5,
            linestyle='--', label='|KF err py|')

    rmse_gps = np.sqrt(np.mean(err_gps_x**2 + err_gps_y**2))
    rmse_kf  = np.sqrt(np.mean(err_kf_x**2  + err_kf_y**2))
    ax.set_title(f'Estimation Error  |  GPS RMSE={rmse_gps:.3f}   KF RMSE={rmse_kf:.3f}')
    ax.set_xlabel('time'); ax.set_ylabel('|error|')
    ax.legend(fontsize=8); ax.grid(True, alpha=0.3)

    fig.tight_layout()
    plt.show()


# ─────────────────────────────────────────────────────────────
# MAIN — run everything
# ─────────────────────────────────────────────────────────────

dt, q, r = 0.1, 0.01, 1.0
T        = 200
seed     = 0

X_true, Y_obs, t = simulate_trajectory(dt=dt, r=r, T=T, seed=seed)
ms, Ps            = run_kalman_filter(Y_obs, dt=dt, q=q, r=r)
plot_results(X_true, Y_obs, ms, Ps, t)
```
