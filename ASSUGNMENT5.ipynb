# Derivation of the Bias in the MLE Covariance Matrix

We begin with the Maximum Likelihood Estimator (MLE) for the population covariance matrix $\Sigma$:

$$
\hat{\Sigma}_{MLE}
=
\frac{1}{n}
\sum_{i=1}^{n}
(X_i-\hat{\mu})
(X_i-\hat{\mu})^T
$$

where

$$
\hat{\mu}
=
\frac{1}{n}
\sum_{i=1}^{n} X_i
$$

is the sample mean.

To determine whether this estimator is biased, write

$$
X_i-\hat{\mu}
=
(X_i-\mu)-(\hat{\mu}-\mu)
$$

Substituting:

$$
\sum_{i=1}^{n}
(X_i-\hat{\mu})(X_i-\hat{\mu})^T
=
\sum_{i=1}^{n}
\left[(X_i-\mu)-(\hat{\mu}-\mu)\right]
\left[(X_i-\mu)-(\hat{\mu}-\mu)\right]^T
$$

Expanding:

$$
=
\sum_{i=1}^{n}(X_i-\mu)(X_i-\mu)^T
-\sum_{i=1}^{n}(X_i-\mu)(\hat{\mu}-\mu)^T
-\sum_{i=1}^{n}(\hat{\mu}-\mu)(X_i-\mu)^T
+\sum_{i=1}^{n}(\hat{\mu}-\mu)(\hat{\mu}-\mu)^T
$$

Since

$$
\sum_{i=1}^{n}(X_i-\mu)
=
n(\hat{\mu}-\mu)
$$

we obtain

$$
=
\sum_{i=1}^{n}(X_i-\mu)(X_i-\mu)^T
-
n(\hat{\mu}-\mu)(\hat{\mu}-\mu)^T
$$

Taking expectations:

$$
E[\hat{\Sigma}_{MLE}]
=
\frac{1}{n}
\sum_{i=1}^{n}
E[(X_i-\mu)(X_i-\mu)^T]
-
E[(\hat{\mu}-\mu)(\hat{\mu}-\mu)^T]
$$

Using

$$
E[(X_i-\mu)(X_i-\mu)^T]=\Sigma
$$

and

$$
E[(\hat{\mu}-\mu)(\hat{\mu}-\mu)^T]
=
Var(\hat{\mu})
=
\frac{\Sigma}{n}
$$

gives

$$
E[\hat{\Sigma}_{MLE}]
=
\Sigma-\frac{\Sigma}{n}
=
\frac{n-1}{n}\Sigma
$$

Since

$$
E[\hat{\Sigma}_{MLE}]
\neq
\Sigma
$$

the MLE covariance estimator is biased downward.

## Bessel's Correction

To remove the bias:

$$
S
=
\frac{n}{n-1}\hat{\Sigma}_{MLE}
$$

Therefore,

$$
S
=
\frac{1}{n-1}
\sum_{i=1}^{n}
(X_i-\hat{\mu})
(X_i-\hat{\mu})^T
$$

Taking expectations:

$$
E[S]
=
\frac{n}{n-1}
E[\hat{\Sigma}_{MLE}]
=
\frac{n}{n-1}
\left(\frac{n-1}{n}\right)\Sigma
=
\Sigma
$$

Thus $S$ is an unbiased estimator of the population covariance matrix.

# Operational Error Diagnostics

### Type I Error ($\alpha$)

A **False Alarm**. The system indicates damage and rejects $H_0$, even though the structure is actually healthy.

### Type II Error ($\beta$)

A **Missed Detection**. The structure is damaged, but the monitoring system fails to reject $H_0$.

# Consequence of an Ultra-Conservative Threshold ($\alpha = 0.0001$)

### 1. Impact on Statistical Power

As $\alpha$ decreases, the rejection threshold becomes much more stringent. Consequently, many real faults are not detected.

The probability of missed detection ($\beta$) increases, and statistical power

$$
1-\beta
$$

decreases significantly.

### 2. Impact on Confidence Ellipsoids

The confidence level becomes

$$
1-\alpha = 0.9999
$$

which enlarges the confidence ellipsoid.

As a result, the healthy-state region expands and includes a much larger range of observations, reducing sensitivity to actual structural damage.

# Slutsky's Theorem

Let $\{X_n\}$ and $\{Y_n\}$ be sequences of random variables (or vectors). If

$$
X_n \xrightarrow{d} X
$$

and

$$
Y_n \xrightarrow{p} c
$$

where $c$ is a constant scalar or matrix, then:

- $X_n + Y_n \xrightarrow{d} X + c$
- $X_nY_n \xrightarrow{d} cX$
- $\dfrac{X_n}{Y_n} \xrightarrow{d} \dfrac{X}{c}$, provided $c \neq 0$

# Operational Distribution Justification

The Lindeberg-Lévy Multivariate Central Limit Theorem (CLT) states that

$$
\sqrt{n}
\left(
\hat{\mu}_n-\mu
\right)
\xrightarrow{d}
\mathcal{N}
\left(
0,\Sigma
\right)
$$

Multiplying by the inverse square root of the population covariance matrix gives

$$
\Sigma^{-1/2}
\left[
\sqrt{n}
\left(
\hat{\mu}_n-\mu
\right)
\right]
\xrightarrow{d}
\mathcal{N}
\left(
0,I
\right)
$$

where $I$ is the identity matrix.

By the Weak Law of Large Numbers (WLLN), the unbiased sample covariance matrix $S$ is a consistent estimator of $\Sigma$. Therefore,

$$
S \xrightarrow{p} \Sigma
$$

and by the Continuous Mapping Theorem,

$$
S^{-1/2}
\xrightarrow{p}
\Sigma^{-1/2}
$$

Applying Slutsky's Theorem, the product of a sequence converging in distribution and a sequence converging in probability yields

$$
S^{-1/2}
\left[
\sqrt{n}
\left(
\hat{\mu}_n-\mu
\right)
\right]
\xrightarrow{d}
\Sigma^{-1/2}
\mathcal{N}(0,\Sigma)
$$

Since

$$
\Sigma^{-1/2}
\Sigma
\Sigma^{-1/2}
=
I
$$

the limiting distribution becomes

$$
\mathcal{N}(0,I)
$$

Hence,

$$
S^{-1/2}
\left[
\sqrt{n}
\left(
\hat{\mu}_n-\mu
\right)
\right]
\xrightarrow{d}
\mathcal{N}(0,I)
$$

Asymptotically, the sample mean therefore satisfies

$$
\sqrt{n}
\left(
\hat{\mu}_n-\mu
\right)
\sim
\mathcal{N}(0,S)
$$

which implies

$$
\hat{\mu}_n
\sim
\mathcal{N}
\left(
\mu,
\frac{1}{n}S
\right)
$$

This result justifies replacing the unknown population covariance matrix $\Sigma$ with its consistent estimator $S$ when constructing asymptotic confidence regions and statistical monitoring boundaries.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# --- DO NOT MODIFY: Synthetic Strain Sensor Data Generator ---
np.random.seed(42)
n_samples = 5000
n_features = 3

base_data = np.random.multivariate_normal(
    mean=[0.5, -0.2, 1.1],
    cov=[[0.09, 0.02, 0.01],
         [0.02, 0.06, 0.03],
         [0.01, 0.03, 0.05]],
    size=n_samples
)

# Injecting structural drift
base_data[4000:, 0] += 0.015
base_data[4000:, 2] -= 0.010

df_strain = pd.DataFrame(
    base_data,
    columns=['Strain_Ch1', 'Strain_Ch2', 'Strain_Ch3']
)
# ----------------------------------------------------------------


def verify_first_moment_homogeneity(df: pd.DataFrame,
                                    g_chunks: int = 5) -> dict:
    """
    Partition the dataset into consecutive chunks and evaluate
    first-moment homogeneity using Wilks' Lambda and Bartlett's
    Chi-Square approximation.
    """

    n = len(df)
    m = df.shape[1]

    if n % g_chunks != 0:
        raise ValueError(
            "Number of samples must be divisible by g_chunks."
        )

    chunk_size = n // g_chunks

    # Global mean vector
    global_mean = df.mean().values.reshape(m, 1)

    # Initialize SSCP matrices
    W = np.zeros((m, m))
    B = np.zeros((m, m))

    # Build W and B
    for i in range(g_chunks):

        chunk = df.iloc[i*chunk_size:(i+1)*chunk_size]
        X = chunk.values

        chunk_mean = chunk.mean().values.reshape(m, 1)

        # Within-group SSCP
        centered = X - chunk_mean.ravel()
        W += centered.T @ centered

        # Between-group SSCP
        diff = chunk_mean - global_mean
        B += chunk_size * (diff @ diff.T)

    # Wilks' Lambda
    wilks_lambda = np.linalg.det(W) / np.linalg.det(W + B)

    # Bartlett approximation
    chi2_calc = -(n - 1 - (m + g_chunks)/2) * np.log(wilks_lambda)

    df_chi2 = m * (g_chunks - 1)

    p_value = 1 - stats.chi2.cdf(chi2_calc, df_chi2)

    conclusion = (
        "Reject H0: Mean vector changed over time "
        "(first moment NOT homogeneous)."
        if p_value < 0.05
        else
        "Fail to reject H0: Mean vector remains constant "
        "(first moment homogeneous)."
    )

    return {
        "Wilks_Lambda": wilks_lambda,
        "Bartlett_ChiSquare": chi2_calc,
        "Degrees_of_Freedom": df_chi2,
        "p_value": p_value,
        "Conclusion": conclusion
    }


# Execute analysis
results = verify_first_moment_homogeneity(df_strain, g_chunks=5)

for k, v in results.items():
    print(f"{k}: {v}")
```

## Testing First-Moment Homogeneity Using MANOVA

To verify the weak i.i.d. assumption, the dataset containing $n=5000$ observations and $m=3$ strain sensor channels was partitioned into $g=5$ consecutive, non-overlapping chunks of size $n_j=1000$.

The within-group SSCP matrix was computed as

$$
\mathbf{W}
=
\sum_{j=1}^{g}
\sum_{i=1}^{n_j}
\left(\mathbf{x}_{ij}-\bar{\mathbf{x}}_j\right)
\left(\mathbf{x}_{ij}-\bar{\mathbf{x}}_j\right)^T.
$$

The between-group SSCP matrix was computed as

$$
\mathbf{B}
=
\sum_{j=1}^{g}
n_j
\left(\bar{\mathbf{x}}_j-\bar{\mathbf{x}}\right)
\left(\bar{\mathbf{x}}_j-\bar{\mathbf{x}}\right)^T.
$$

where $\bar{\mathbf{x}}_j$ denotes the mean vector of chunk $j$ and $\bar{\mathbf{x}}$ denotes the overall mean vector.

Wilks' Lambda was calculated using

$$
\Lambda
=
\frac{|\mathbf{W}|}
{|\mathbf{B}+\mathbf{W}|}.
$$

To evaluate statistical significance, Bartlett's chi-square approximation was applied:

$$
\chi^2_{\mathrm{calc}}
=
-\left(
n-1-\frac{m+g}{2}
\right)
\ln(\Lambda).
$$

The computed results were:

- Wilks' Lambda: $\Lambda = 0.990452$
- Bartlett Chi-Square: $\chi^2 = 47.922$
- Degrees of Freedom: $df = 12$
- p-value: $p = 3.23 \times 10^{-6}$

### Hypothesis Test

**Null Hypothesis ($H_0$):** The multivariate mean vector remains constant across all five chunks.

**Alternative Hypothesis ($H_1$):** At least one chunk has a different mean vector.

Using a significance level of $\alpha = 0.05$,

$$
p = 3.23 \times 10^{-6} < 0.05.
$$

Therefore, the null hypothesis is rejected.

### Conclusion

The MANOVA analysis reveals a statistically significant difference in the multivariate mean vector across the five consecutive time blocks. Therefore, the assumption of constant joint mean (first-moment homogeneity) is not satisfied.

The baseline center of the process has shifted over time, indicating the presence of structural drift in the sensor measurements. This result is consistent with the artificial drift introduced into Sensor Channel 1 and Sensor Channel 3 during the final 1000 observations.

# **Geometric Subspace Optimization via Principal Component Analysis (PCA)**

**PART A**

# 1. Covariance Structure in Principal Component Space

Let

$$
\widetilde{\mathbf{X}}_i
=
(\mathbf{X}_i-\boldsymbol{\mu})
\in \mathbb{R}^{m}
$$

be a centered random vector with covariance matrix

$$
\boldsymbol{\Sigma}
=
\mathbb{E}
\left[
\widetilde{\mathbf{X}}_i
\widetilde{\mathbf{X}}_i^T
\right].
$$

Assume the spectral decomposition

$$
\boldsymbol{\Sigma}
=
\mathbf{P}
\mathbf{\Lambda}
\mathbf{P}^T
$$

where

$$
\mathbf{\Lambda}
=
\operatorname{diag}
(\lambda_1,\lambda_2,\ldots,\lambda_m)
$$

and $\mathbf{P}$ is orthogonal:

$$
\mathbf{P}^T\mathbf{P}
=
\mathbf{P}\mathbf{P}^T
=
\mathbf{I}.
$$

Define the transformed coordinates

$$
\mathbf{Z}_i
=
\mathbf{P}^T
\widetilde{\mathbf{X}}_i.
$$

## Proof of the Covariance in Principal Space

The covariance matrix of $\mathbf{Z}_i$ is

$$
\mathbb{E}
\left[
\mathbf{Z}_i
\mathbf{Z}_i^T
\right].
$$

Substituting the definition of $\mathbf{Z}_i$:

$$
=
\mathbb{E}
\left[
(\mathbf{P}^T\widetilde{\mathbf{X}}_i)
(\mathbf{P}^T\widetilde{\mathbf{X}}_i)^T
\right].
$$

Using

$$
(\mathbf{AB})^T
=
\mathbf{B}^T\mathbf{A}^T,
$$

we obtain

$$
=
\mathbb{E}
\left[
\mathbf{P}^T
\widetilde{\mathbf{X}}_i
\widetilde{\mathbf{X}}_i^T
\mathbf{P}
\right].
$$

Since $\mathbf{P}$ is deterministic,

$$
=
\mathbf{P}^T
\mathbb{E}
\left[
\widetilde{\mathbf{X}}_i
\widetilde{\mathbf{X}}_i^T
\right]
\mathbf{P}.
$$

Using the definition of covariance,

$$
=
\mathbf{P}^T
\boldsymbol{\Sigma}
\mathbf{P}.
$$

Substituting

$$
\boldsymbol{\Sigma}
=
\mathbf{P}
\mathbf{\Lambda}
\mathbf{P}^T,
$$

gives

$$
=
\mathbf{P}^T
(\mathbf{P}\mathbf{\Lambda}\mathbf{P}^T)
\mathbf{P}.
$$

Using orthogonality,

$$
\mathbf{P}^T\mathbf{P}
=
\mathbf{I},
$$

therefore

$$
=
\mathbf{\Lambda}.
$$

Hence

$$
\boxed{
\mathbb{E}
\left[
\mathbf{Z}_i
\mathbf{Z}_i^T
\right]
=
\mathbf{\Lambda}
}
$$

which is diagonal.

## Geometric and Statistical Interpretation

Since

$$
\mathbf{\Lambda}
=
\operatorname{diag}
(\lambda_1,\ldots,\lambda_m),
$$

all off-diagonal entries are zero:

$$
\operatorname{Cov}
(Z_{i,j},Z_{i,k})
=
0,
\qquad
j\neq k.
$$

Therefore:

- Principal components are mutually uncorrelated.
- Each coordinate measures variability along one eigenvector direction.
- Variance along component $j$ equals $\lambda_j$.
- No linear redundancy exists between different principal coordinates.

Geometrically, PCA rotates the coordinate system so that the covariance ellipsoid aligns exactly with the coordinate axes.

---

# 2. Preservation of Total Variance

The total variance in the original space equals

$$
\operatorname{tr}(\boldsymbol{\Sigma}).
$$

Using

$$
\boldsymbol{\Sigma}
=
\mathbf{P}
\mathbf{\Lambda}
\mathbf{P}^T,
$$

and the cyclic property of trace,

$$
\operatorname{tr}(ABC)
=
\operatorname{tr}(BCA),
$$

we obtain

$$
\operatorname{tr}(\boldsymbol{\Sigma})
=
\operatorname{tr}
(\mathbf{P}\mathbf{\Lambda}\mathbf{P}^T).
$$

Applying cyclic permutation,

$$
=
\operatorname{tr}
(\mathbf{\Lambda}\mathbf{P}^T\mathbf{P}).
$$

Using

$$
\mathbf{P}^T\mathbf{P}
=
\mathbf{I},
$$

gives

$$
=
\operatorname{tr}(\mathbf{\Lambda}).
$$

Since the trace of a diagonal matrix equals the sum of its diagonal entries,

$$
\operatorname{tr}(\mathbf{\Lambda})
=
\sum_{j=1}^{m}\lambda_j.
$$

Therefore

$$
\boxed{
\operatorname{tr}(\boldsymbol{\Sigma})
=
\operatorname{tr}(\mathbf{\Lambda})
=
\sum_{j=1}^{m}\lambda_j
}
$$

showing that PCA preserves total variance.

## Cumulative Explained Variance Ratio

For the first $k$ principal components,

$$
\boxed{
\Phi(k)
=
\frac{\sum_{j=1}^{k}\lambda_j}
{\sum_{j=1}^{m}\lambda_j}
}
$$

## Residual Unexplained Variance Ratio

For the discarded subspace,

$$
\boxed{
\Psi(k)
=
\frac{\sum_{j=k+1}^{m}\lambda_j}
{\sum_{j=1}^{m}\lambda_j}
}
$$

and

$$
\Phi(k)+\Psi(k)=1.
$$

# 3. Reconstruction and Residual Analysis in SHM

Assume

$$
(\mathbf{x}_i-\hat{\boldsymbol{\mu}}_n)
=
\hat{\mathbf{P}}_k\mathbf{z}_{i,k}
+
\hat{\mathbf{P}}_{m-k}\mathbf{z}_{i,m-k}.
$$

## Reconstruction Vector

The PCA reconstruction using the retained subspace is

$$
\boxed{
\hat{\mathbf{x}}_i
=
\hat{\boldsymbol{\mu}}_n
+
\hat{\mathbf{P}}_k
\mathbf{z}_{i,k}
}
$$

## Residual Error Vector

The reconstruction error is

$$
\boxed{
\mathbf{e}_i
=
\mathbf{x}_i-\hat{\mathbf{x}}_i
=
\hat{\mathbf{P}}_{m-k}
\mathbf{z}_{i,m-k}
}
$$

## Proof of Energy Decomposition

The squared norm is

$$
\|
\mathbf{x}_i-\hat{\boldsymbol{\mu}}_n
\|^2.
$$

Substituting the decomposition,

$$
=
\|
\hat{\mathbf{P}}_k\mathbf{z}_{i,k}
+
\hat{\mathbf{P}}_{m-k}\mathbf{z}_{i,m-k}
\|^2.
$$

Expanding,

$$
=
\|
\hat{\mathbf{P}}_k\mathbf{z}_{i,k}
\|^2
+
\|
\hat{\mathbf{P}}_{m-k}\mathbf{z}_{i,m-k}
\|^2
+
2
(\hat{\mathbf{P}}_k\mathbf{z}_{i,k})^T
(\hat{\mathbf{P}}_{m-k}\mathbf{z}_{i,m-k}).
$$

Because the two subspaces are orthogonal,

$$
\hat{\mathbf{P}}_k^T
\hat{\mathbf{P}}_{m-k}
=
0,
$$

the cross term vanishes.

Also,

$$
\hat{\mathbf{P}}_k^T
\hat{\mathbf{P}}_k
=
I
$$

and

$$
\hat{\mathbf{P}}_{m-k}^T
\hat{\mathbf{P}}_{m-k}
=
I.
$$

Hence

$$
\|
\hat{\mathbf{P}}_k\mathbf{z}_{i,k}
\|^2
=
\|
\mathbf{z}_{i,k}
\|^2
$$

and

$$
\|
\hat{\mathbf{P}}_{m-k}\mathbf{z}_{i,m-k}
\|^2
=
\|
\mathbf{z}_{i,m-k}
\|^2.
$$

Therefore

$$
\boxed{
\|
\mathbf{x}_i-\hat{\boldsymbol{\mu}}_n
\|^2
=
\|
\mathbf{z}_{i,k}
\|^2
+
\|
\mathbf{z}_{i,m-k}
\|^2
}
$$

which is the Pythagorean decomposition of energy.

# Hotelling's $T^2$ vs Residual $Q$ Statistic

## Hotelling's $T^2$

Measures variation inside the retained PCA subspace:

$$
T^2
=
\mathbf{z}_{i,k}^T
\mathbf{\Lambda}_k^{-1}
\mathbf{z}_{i,k}.
$$

Sensitive to:

- Environmental changes
- Operating-condition variations
- Wind loading
- Temperature effects
- Changes already represented by the PCA model

Therefore an extreme operational wind-load event is most likely detected by

$$
\boxed{T^2}
$$

because it causes unusually large scores within the principal subspace.

## Residual $Q$ Statistic (SPE)

Measures variation outside the retained PCA model:

$$
Q
=
\|
\mathbf{e}_i
\|^2.
$$

Sensitive to:

- Structural damage
- Cracks
- Delamination
- Sensor faults
- New dynamic behaviors not present during training

Therefore a fatigue crack altering structural load paths is most likely detected by

$$
\boxed{Q}
$$

because the crack introduces behavior not represented by the learned principal subspace.

### Summary

| Event | Most Sensitive Statistic |
|---------|---------|
| Extreme wind/environmental loading | $T^2$ |
| Fatigue crack / structural fracture | $Q$ (SPE) |

## **Part B: Numerical Verification**



$$
\textbf{Setup}
$$

$$
\text{Baseline archive: } n = 3000 \text{ nominal snapshots}, \quad m = 4 \text{ sensors}
$$

$$
\text{Each snapshot: } \mathbf{x}_i \in \mathbb{R}^4, \quad i = 1, \dots, n
$$

---

$$
\textbf{Step 1 — Standardise and Centre}
$$

$$
\hat{\mu}_n = \frac{1}{n}\sum_{i=1}^n \mathbf{x}_i \in \mathbb{R}^m, \qquad
\hat{\sigma}_j = \sqrt{\frac{1}{n-1}\sum_{i=1}^n (x_{i,j} - \hat{\mu}_j)^2}
$$

$$
\tilde{x}_{i,j} = \frac{x_{i,j} - \hat{\mu}_j}{\hat{\sigma}_j}
\qquad \Longrightarrow \qquad
\tilde{\mathbf{X}} \in \mathbb{R}^{n \times m}
$$

---

$$
\textbf{Step 2 — Unbiased Sample Covariance Matrix and Spectral Decomposition}
$$

$$
\mathbf{S} = \frac{1}{n-1}\sum_{i=1}^n (\tilde{\mathbf{x}}_i - \hat{\boldsymbol{\mu}})(\tilde{\mathbf{x}}_i - \hat{\boldsymbol{\mu}})^\top \in \mathbb{R}^{m \times m}
$$

$$
\text{By the Spectral Theorem (}\mathbf{S}\text{ symmetric PSD):}
\qquad \mathbf{S} = \hat{\mathbf{P}}\,\hat{\boldsymbol{\Lambda}}\,\hat{\mathbf{P}}^\top
$$

$$
\hat{\mathbf{P}} = [\hat{\mathbf{p}}_1, \dots, \hat{\mathbf{p}}_m] \in \mathbb{R}^{m \times m}, \quad \hat{\mathbf{P}}^\top\hat{\mathbf{P}} = \mathbf{I}
$$

$$
\hat{\boldsymbol{\Lambda}} = \mathrm{diag}(\hat{\lambda}_1, \dots, \hat{\lambda}_m), \quad \hat{\lambda}_1 \ge \hat{\lambda}_2 \ge \cdots \ge \hat{\lambda}_m \ge 0
$$

---

$$
\textbf{Step 3 — PC Scores and Subspace Partition for each } k
$$

$$
\mathbf{z}_i = \hat{\mathbf{P}}^\top \tilde{\mathbf{x}}_i \in \mathbb{R}^m
\qquad \text{(full projection)}
$$

$$
\mathbf{z}_{i,k} = [z_{i,1}, \dots, z_{i,k}]^\top \in \mathbb{R}^k
\qquad \text{(principal subspace scores)}
$$

$$
\mathbf{z}_{i,m-k} = [z_{i,k+1}, \dots, z_{i,m}]^\top \in \mathbb{R}^{m-k}
\qquad \text{(residual subspace scores)}
$$

**Hotelling's $T^2$ statistic** — Mahalanobis distance in the principal subspace:

$$
T^2_i = \mathbf{z}_{i,k}^\top \hat{\boldsymbol{\Lambda}}_k^{-1} \mathbf{z}_{i,k}
= \sum_{j=1}^{k} \frac{z_{i,j}^2}{\hat{\lambda}_j}
$$

**$Q$ statistic (Squared Prediction Error)** — residual energy:

$$
Q_i = \|\mathbf{e}_i\|^2 = \mathbf{z}_{i,m-k}^\top \mathbf{z}_{i,m-k}
= \sum_{j=k+1}^{m} z_{i,j}^2
$$

---

$$
\textbf{Step 4 — Empirical Expected Values}
$$

$$
\mathbb{E}[T^2] \approx \frac{1}{n}\sum_{i=1}^n T^2_i
\xrightarrow{\;n \to \infty\;} k
\qquad \text{(converges to degrees of freedom under i.i.d. normality)}
$$

$$
\mathbb{E}[Q] \approx \frac{1}{n}\sum_{i=1}^n Q_i
\xrightarrow{\;n \to \infty\;} \sum_{j=k+1}^{m} \hat{\lambda}_j
\qquad \text{(residual variance in discarded axes)}
$$

**Cumulative explained variance ratio:**

$$
\Phi(k) = \frac{\sum_{j=1}^{k}\hat{\lambda}_j}{\sum_{j=1}^{m}\hat{\lambda}_j}, \qquad
\Psi(k) = 1 - \Phi(k) = \frac{\sum_{j=k+1}^{m}\hat{\lambda}_j}{\sum_{j=1}^{m}\hat{\lambda}_j}
$$

| $k$ | $\mathbb{E}[T^2]$ | $\mathbb{E}[Q]$ | $\Phi(k)$ | Diagnostic focus |
|---|---|---|---|---|
| 1 | $\approx 1$ | large | low | Catches only dominant mode shifts |
| 2 | $\approx 2$ | medium | medium | Balanced coverage |
| 3 | $\approx 3$ | $\approx \hat{\lambda}_4$ | high | Near-full structure; Q very sensitive |

```python
import numpy as np

# ── Synthetic baseline data (m=4 sensors, n=3000 snapshots) ──────────────
np.random.seed(42)
n, m = 3000, 4
true_cov = np.array([
    [3.0, 1.5, 0.8, 0.2],
    [1.5, 2.5, 0.6, 0.1],
    [0.8, 0.6, 1.5, 0.3],
    [0.2, 0.1, 0.3, 0.8],
])
X = np.random.multivariate_normal(np.zeros(m), true_cov, size=n)

# ── Replace the block above with your real data ───────────────────────────
# X = inspector.df[column_names].values   ← if using DataInspector

class DiagnosticsLayer:
    def compute_pca_diagnostics(self, X: np.ndarray) -> dict:
        """
        Compute Hotelling T² and Q statistics for k ∈ {1, 2, …, m-1}.

        Parameters
        ----------
        X : np.ndarray, shape (n, m)
            Raw baseline snapshot matrix.

        Returns
        -------
        dict with keys 'k1', 'k2', 'k3', each mapping to:
            T2_samples    (n,)  — per-sample Hotelling T²
            Q_samples     (n,)  — per-sample Q / SPE
            E_T2          scalar — empirical mean of T²
            E_Q           scalar — empirical mean of Q
            var_explained scalar — cumulative explained variance Φ(k)
        """
        n, m = X.shape

        # ── Step 1: Standardise and centre ──────────────────────────────
        mu_hat    = X.mean(axis=0)           # (m,)  empirical mean
        sigma_hat = X.std(axis=0, ddof=1)    # (m,)  empirical std  (Bessel-corrected)
        X_std     = (X - mu_hat) / sigma_hat # (n, m) zero-mean, unit-variance

        # ── Step 2: Unbiased sample covariance S = 1/(n-1) X̃ᵀX̃ ────────
        S = np.cov(X_std, rowvar=False, ddof=1)          # (m, m)

        # Spectral decomposition via eigh (symmetric matrix → real eigenvalues)
        eigenvalues, eigenvectors = np.linalg.eigh(S)

        # Sort descending: λ̂₁ ≥ λ̂₂ ≥ … ≥ λ̂ₘ
        idx        = np.argsort(eigenvalues)[::-1]
        lambda_hat = eigenvalues[idx]      # (m,)
        P_hat      = eigenvectors[:, idx]  # (m, m)
        total_var  = lambda_hat.sum()

        # Project all samples to full principal space once: Z = X̃ P̂
        Z = X_std @ P_hat   # (n, m)

        results = {}

        # ── Steps 3 & 4: Loop over k ∈ {1, 2, 3} ───────────────────────
        for k in range(1, m):   # m=4 → k ∈ {1, 2, 3}

            Z_k   = Z[:, :k]   # (n, k)   principal subspace scores
            Z_res = Z[:, k:]   # (n, m-k) residual subspace scores

            # Hotelling T²_i = Σⱼ₌₁ᵏ  z²ᵢⱼ / λ̂ⱼ
            T2_samples = (Z_k ** 2 / lambda_hat[:k]).sum(axis=1)   # (n,)

            # Q_i = Σⱼ₌ₖ₊₁ᵐ z²ᵢⱼ
            Q_samples  = (Z_res ** 2).sum(axis=1)                   # (n,)

            E_T2          = T2_samples.mean()
            E_Q           = Q_samples.mean()
            var_explained = lambda_hat[:k].sum() / total_var        # Φ(k)

            results[f'k{k}'] = {
                'T2_samples'    : T2_samples,
                'Q_samples'     : Q_samples,
                'E_T2'          : E_T2,
                'E_Q'           : E_Q,
                'var_explained' : var_explained,
            }

        return results, lambda_hat


# ── Run and print results ─────────────────────────────────────────────────
diag = DiagnosticsLayer()
res, lambda_hat = diag.compute_pca_diagnostics(X)

print(f"\nEigenvalues: {np.round(lambda_hat, 4)}\n")
print(f"{'k':>3}  {'E[T²]':>10}  {'E[Q]':>10}  {'Φ(k) — var explained':>22}  {'Ψ(k) — residual var':>22}")
print("-" * 75)
for k in range(1, m):
    r = res[f'k{k}']
    psi = 1 - r['var_explained']
    print(f"{k:>3}  {r['E_T2']:>10.4f}  {r['E_Q']:>10.4f}  {r['var_explained']:>21.2%}  {psi:>21.2%}")

print("\nVerification checks:")
for k in range(1, m):
    r = res[f'k{k}']
    resid_var = lambda_hat[k:].sum()
    print(f"  k={k}: E[T²] ≈ {r['E_T2']:.4f}  (theory → {k})  |  "
          f"E[Q] ≈ {r['E_Q']:.4f}  (theory → Σλ̂_residual = {resid_var:.4f})")
```

# **Latent Subspace Decomposition via Factor Analysis (FA)**

**PART A**

## Question 1 — Fundamental Equation of Factor Analysis

### Part A — Proof

Given the generative model:
$$\mathbf{Z}_i = \boldsymbol{\Lambda}\mathbf{F}_i + \boldsymbol{\epsilon}_i$$

We compute $\mathbf{R} = \mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^\top]$ by expanding the outer product:

$$
\mathbf{R}
= \mathbb{E}\!\left[(\boldsymbol{\Lambda}\mathbf{F}_i + \boldsymbol{\epsilon}_i)
  (\boldsymbol{\Lambda}\mathbf{F}_i + \boldsymbol{\epsilon}_i)^\top\right]
$$

$$
= \mathbb{E}\!\left[\boldsymbol{\Lambda}\mathbf{F}_i\mathbf{F}_i^\top\boldsymbol{\Lambda}^\top\right]
+ \mathbb{E}\!\left[\boldsymbol{\Lambda}\mathbf{F}_i\boldsymbol{\epsilon}_i^\top\right]
+ \mathbb{E}\!\left[\boldsymbol{\epsilon}_i\mathbf{F}_i^\top\boldsymbol{\Lambda}^\top\right]
+ \mathbb{E}\!\left[\boldsymbol{\epsilon}_i\boldsymbol{\epsilon}_i^\top\right]
$$

Applying the three population constraints:

$$
\mathbb{E}[\mathbf{F}_i\mathbf{F}_i^\top] = \mathbf{I}_{k\times k}, \qquad
\mathbb{E}[\boldsymbol{\epsilon}_i\mathbf{F}_i^\top] = \mathbf{0}_{m\times k}, \qquad
\mathbb{E}[\boldsymbol{\epsilon}_i\boldsymbol{\epsilon}_i^\top] = \boldsymbol{\Psi}
$$

Each term reduces:

$$
\mathbb{E}\!\left[\boldsymbol{\Lambda}\mathbf{F}_i\mathbf{F}_i^\top\boldsymbol{\Lambda}^\top\right]
= \boldsymbol{\Lambda}\,\mathbb{E}[\mathbf{F}_i\mathbf{F}_i^\top]\,\boldsymbol{\Lambda}^\top
= \boldsymbol{\Lambda}\mathbf{I}\boldsymbol{\Lambda}^\top
= \boldsymbol{\Lambda}\boldsymbol{\Lambda}^\top
$$

$$
\mathbb{E}\!\left[\boldsymbol{\Lambda}\mathbf{F}_i\boldsymbol{\epsilon}_i^\top\right]
= \boldsymbol{\Lambda}\,\mathbb{E}[\mathbf{F}_i\boldsymbol{\epsilon}_i^\top]
= \boldsymbol{\Lambda}\,\bigl(\mathbb{E}[\boldsymbol{\epsilon}_i\mathbf{F}_i^\top]\bigr)^\top
= \boldsymbol{\Lambda}\,\mathbf{0}^\top = \mathbf{0}
$$

$$
\mathbb{E}\!\left[\boldsymbol{\epsilon}_i\mathbf{F}_i^\top\boldsymbol{\Lambda}^\top\right]
= \mathbb{E}[\boldsymbol{\epsilon}_i\mathbf{F}_i^\top]\,\boldsymbol{\Lambda}^\top
= \mathbf{0}\,\boldsymbol{\Lambda}^\top = \mathbf{0}
$$

Substituting back:

$$
\boxed{\mathbf{R} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^\top + \boldsymbol{\Psi}}
$$

This is the **Fundamental Equation of Factor Analysis**. It decomposes the total sensor correlation structure into a low-rank shared component $\boldsymbol{\Lambda}\boldsymbol{\Lambda}^\top$ driven by the $k$ common latent factors, and a diagonal noise floor $\boldsymbol{\Psi}$ unique to each sensor channel. $\blacksquare$

---

### Part B — Communality and Uniqueness

Since $\mathbf{Z}_i$ is normalized, every diagonal entry of $\mathbf{R}$ equals unity:

$$R_{jj} = \mathbb{E}[Z_{i,j}^2] = 1$$

The $j$-th diagonal of $\boldsymbol{\Lambda}\boldsymbol{\Lambda}^\top$ is the squared Euclidean row-norm of the $j$-th row of $\boldsymbol{\Lambda}$:

$$\left(\boldsymbol{\Lambda}\boldsymbol{\Lambda}^\top\right)_{jj} = \sum_{r=1}^{k}\lambda_{j,r}^2$$

Therefore the $j$-th diagonal entry of the fundamental equation gives:

$$1 = \sum_{r=1}^{k}\lambda_{j,r}^2 + \varphi_j^2$$

**Definitions:**

$$\boxed{h_j^2 \;\triangleq\; \sum_{r=1}^{k}\lambda_{j,r}^2} \qquad \textbf{Communality of sensor } j$$

$$\boxed{\varphi_j^2 \;\triangleq\; 1 - h_j^2} \qquad \textbf{Uniqueness of sensor } j$$

with the unit-variance invariant: $h_j^2 + \varphi_j^2 = 1 \quad \forall\, j$.

**Physical interpretations in an asset diagnostics system:**

**Communality $h_j^2$** is the proportion of sensor $j$'s total variance that is collectively explained by the $k$ shared latent structural modes (e.g.\ global vibration, thermal gradient, mechanical load path). A sensor with $h_j^2 = 0.92$ contributes $92\%$ of its signal energy to the shared correlation network. High communality confirms the sensor is genuinely coupled to the system's structural dynamics.

**Uniqueness $\varphi_j^2$** is the proportion of sensor $j$'s variance that is statistically isolated — driven by local measurement noise, sensor drift, or a localized physical anomaly not shared by any other sensor. Under nominal conditions $\varphi_j^2$ is small and stable. A sudden increase in $\varphi_j^2$ toward $1.0$ is a direct diagnostic flag: sensor $j$ has decoupled from the shared structural network, indicating either sensor degradation or an incipient localized fault that has not yet propagated to the global structural modes.

---

## Question 2 — Varimax Rotation and Interpretability

### Part A — Raw PCA Loading vs Varimax-Rotated Loading

A **raw principal component loading vector** $\hat{\mathbf{p}}_r \in \mathbb{R}^m$ is the $r$-th eigenvector of the sample covariance matrix $\mathbf{S}$, defined purely by the maximum variance criterion: each successive component captures the largest possible residual variance subject to orthogonality with all prior components. As a consequence, every sensor typically receives a non-negligible loading on every component — the explained variance is spread diffusely across all $m$ entries of $\hat{\mathbf{p}}_r$. No individual component can be attributed to a specific physical mechanism, making structural interpretation difficult.

A **Varimax-rotated factor loading vector** is obtained by applying an orthogonal rotation matrix $\mathbf{T} \in \mathbb{R}^{k\times k}$ (with $\mathbf{T}^\top\mathbf{T} = \mathbf{I}$) to the initial loading matrix:

$$\boldsymbol{\Lambda}^* = \boldsymbol{\Lambda}\,\mathbf{T}$$

The Varimax criterion selects $\mathbf{T}$ to maximize the variance of the squared loadings within each factor column:

$$
\mathbf{T}^* = \underset{\mathbf{T}:\,\mathbf{T}^\top\mathbf{T}=\mathbf{I}}{\arg\max}
\sum_{r=1}^{k}\left[
\frac{1}{m}\sum_{j=1}^{m}\lambda_{j,r}^{*4}
- \left(\frac{1}{m}\sum_{j=1}^{m}\lambda_{j,r}^{*2}\right)^{\!2}
\right]
$$

This drives each rotated column toward a **simple structure**: each factor loads heavily on a small subset of sensors and near-zero on the rest. In a diagnostics system, one rotated factor might load exclusively on temperature and cooling-flow sensors (thermal mode) while another loads exclusively on vibration and RPM sensors (mechanical mode) — a physically interpretable partition that is impossible to read from the raw eigenvectors.

---

### Part B — Why Orthogonal Rotation Preserves Communality and the Global Approximation

**Preservation of communality $h_j^2$:**

The communality of sensor $j$ equals the squared Euclidean norm of the $j$-th row $\boldsymbol{\lambda}_j^\top \in \mathbb{R}^k$ of $\boldsymbol{\Lambda}$:

$$h_j^2 = \|\boldsymbol{\lambda}_j\|^2 = \boldsymbol{\lambda}_j^\top\boldsymbol{\lambda}_j$$

Under the rotation $\boldsymbol{\Lambda}^* = \boldsymbol{\Lambda}\mathbf{T}$, the $j$-th row transforms as $\boldsymbol{\lambda}_j^{*\top} = \boldsymbol{\lambda}_j^\top\mathbf{T}$. Therefore:

$$
h_j^{*2}
= \|\boldsymbol{\lambda}_j^*\|^2
= \boldsymbol{\lambda}_j^\top\mathbf{T}\,\mathbf{T}^\top\boldsymbol{\lambda}_j
= \boldsymbol{\lambda}_j^\top\mathbf{I}\,\boldsymbol{\lambda}_j
= \boldsymbol{\lambda}_j^\top\boldsymbol{\lambda}_j
= h_j^2
$$

$$\therefore \quad h_j^{*2} = h_j^2 \quad \forall\, j \in \{1,\dots,m\} \qquad \blacksquare$$

**Preservation of the global covariance approximation:**

$$
\boldsymbol{\Lambda}^*{\boldsymbol{\Lambda}^*}^\top
= (\boldsymbol{\Lambda}\mathbf{T})(\boldsymbol{\Lambda}\mathbf{T})^\top
= \boldsymbol{\Lambda}\,\mathbf{T}\mathbf{T}^\top\,\boldsymbol{\Lambda}^\top
= \boldsymbol{\Lambda}\,\mathbf{I}\,\boldsymbol{\Lambda}^\top
= \boldsymbol{\Lambda}\boldsymbol{\Lambda}^\top
$$

Therefore:

$$\boldsymbol{\Lambda}^*{\boldsymbol{\Lambda}^*}^\top + \boldsymbol{\Psi} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^\top + \boldsymbol{\Psi} = \mathbf{R} \qquad \blacksquare$$

**Geometric summary:** An orthogonal rotation $\mathbf{T}$ is a rigid transformation of the $k$-dimensional factor subspace. It redistributes explained variance among the $k$ columns — changing which sensors load on which factor — but cannot add or remove variance from the shared subspace $\boldsymbol{\Lambda}\boldsymbol{\Lambda}^\top$. The subspace itself is invariant; only the coordinate axes within it are reoriented. Communality, being the orthogonal projection of each sensor onto this fixed subspace, is therefore geometrically unchanged regardless of how the axes are rotated.

## Question 3 — Thomson's Regression Method for Factor Score Estimation

### Part A — Joint Covariance Matrix $\boldsymbol{\Sigma}_{YY}$

Construct the partitioned joint vector:

$$\mathbf{Y}_i = \begin{bmatrix} \mathbf{Z}_i \\ \mathbf{F}_i \end{bmatrix} \in \mathbb{R}^{m+k}$$

Since $\mathbb{E}[\mathbf{Z}_i] = \mathbf{0}$ and $\mathbb{E}[\mathbf{F}_i] = \mathbf{0}$, the joint mean is zero and the joint covariance expands as:

$$
\boldsymbol{\Sigma}_{YY}
= \mathbb{E}[\mathbf{Y}_i\mathbf{Y}_i^\top]
= \mathbb{E}\!\left[\begin{bmatrix}\mathbf{Z}_i\\\mathbf{F}_i\end{bmatrix}
\begin{bmatrix}\mathbf{Z}_i^\top & \mathbf{F}_i^\top\end{bmatrix}\right]
= \begin{bmatrix}
\mathbb{E}[\mathbf{Z}_i\mathbf{Z}_i^\top] & \mathbb{E}[\mathbf{Z}_i\mathbf{F}_i^\top] \\
\mathbb{E}[\mathbf{F}_i\mathbf{Z}_i^\top] & \mathbb{E}[\mathbf{F}_i\mathbf{F}_i^\top]
\end{bmatrix}
$$

We evaluate each block using the fundamental constraints:

**Block (1,1):** By the fundamental equation already proved:
$$\mathbb{E}[\mathbf{Z}_i\mathbf{Z}_i^\top] = \mathbf{R}$$

**Block (1,2):** Substitute $\mathbf{Z}_i = \boldsymbol{\Lambda}\mathbf{F}_i + \boldsymbol{\epsilon}_i$:
$$
\mathbb{E}[\mathbf{Z}_i\mathbf{F}_i^\top]
= \mathbb{E}\!\left[(\boldsymbol{\Lambda}\mathbf{F}_i + \boldsymbol{\epsilon}_i)\mathbf{F}_i^\top\right]
= \boldsymbol{\Lambda}\,\mathbb{E}[\mathbf{F}_i\mathbf{F}_i^\top] + \mathbb{E}[\boldsymbol{\epsilon}_i\mathbf{F}_i^\top]
= \boldsymbol{\Lambda}\mathbf{I} + \mathbf{0}
= \boldsymbol{\Lambda}
$$

**Block (2,1):** By symmetry of expectation:
$$\mathbb{E}[\mathbf{F}_i\mathbf{Z}_i^\top] = \bigl(\mathbb{E}[\mathbf{Z}_i\mathbf{F}_i^\top]\bigr)^\top = \boldsymbol{\Lambda}^\top$$

**Block (2,2):** By the orthogonal factor constraint:
$$\mathbb{E}[\mathbf{F}_i\mathbf{F}_i^\top] = \mathbf{I}_{k\times k}$$

Assembling all four blocks:

$$
\boxed{
\boldsymbol{\Sigma}_{YY}
= \begin{bmatrix} \mathbf{R} & \boldsymbol{\Lambda} \\ \boldsymbol{\Lambda}^\top & \mathbf{I}_{k\times k} \end{bmatrix}
} \qquad \blacksquare
$$

---

### Part B — Derivation of Thomson's Estimator

Under the Multivariate Normal assumption, the conditional mean of a jointly Gaussian partitioned system

$$\begin{bmatrix}\mathbf{X}_1\\\mathbf{X}_2\end{bmatrix} \sim \mathcal{N}\!\left(\begin{bmatrix}\boldsymbol{\mu}_1\\\boldsymbol{\mu}_2\end{bmatrix},\begin{bmatrix}\boldsymbol{\Sigma}_{11}&\boldsymbol{\Sigma}_{12}\\\boldsymbol{\Sigma}_{21}&\boldsymbol{\Sigma}_{22}\end{bmatrix}\right)$$

is given by the linear projection identity:

$$\mathbb{E}[\mathbf{X}_2\mid\mathbf{x}_1] = \boldsymbol{\mu}_2 + \boldsymbol{\Sigma}_{21}\boldsymbol{\Sigma}_{11}^{-1}(\mathbf{x}_1 - \boldsymbol{\mu}_1)$$

We map the partitioned system to our factor analysis setup by identifying:

$$
\mathbf{X}_1 \leftarrow \mathbf{Z}_i, \qquad \mathbf{X}_2 \leftarrow \mathbf{F}_i
$$

$$
\boldsymbol{\mu}_1 = \mathbb{E}[\mathbf{Z}_i] = \mathbf{0}, \qquad \boldsymbol{\mu}_2 = \mathbb{E}[\mathbf{F}_i] = \mathbf{0}
$$

Reading the corresponding blocks directly from $\boldsymbol{\Sigma}_{YY}$:

$$
\boldsymbol{\Sigma}_{11} = \mathbb{E}[\mathbf{Z}_i\mathbf{Z}_i^\top] = \mathbf{R}, \qquad
\boldsymbol{\Sigma}_{21} = \mathbb{E}[\mathbf{F}_i\mathbf{Z}_i^\top] = \boldsymbol{\Lambda}^\top
$$

Substituting into the projection identity:

$$
\mathbf{f}_i
= \mathbb{E}[\mathbf{F}_i\mid\mathbf{z}_i]
= \mathbf{0} + \boldsymbol{\Lambda}^\top\mathbf{R}^{-1}(\mathbf{z}_i - \mathbf{0})
$$

$$
\boxed{\mathbf{f}_i = \boldsymbol{\Lambda}^\top\mathbf{R}^{-1}\mathbf{z}_i} \qquad \blacksquare
$$

This is **Thomson's Regression Estimator** for the latent factor scores. It is the **Best Linear Unbiased Predictor (BLUP)** of $\mathbf{F}_i$ given the observation $\mathbf{z}_i$: it minimizes the mean squared error $\mathbb{E}[\|\mathbf{F}_i - \mathbf{f}_i\|^2]$ over all linear estimators, and remains optimal even when the data deviates moderately from strict normality.

**Structural interpretation:** The term $\mathbf{R}^{-1}\mathbf{z}_i$ decorrelates the observed sensor readings by inverting the full correlation structure. The subsequent left-multiplication by $\boldsymbol{\Lambda}^\top$ then projects this whitened observation onto the $k$-dimensional factor subspace, extracting the posterior mean coordinates of the hidden physical state vector given the observed sensor snapshot.

## **Part B: Parametric Factor Partition Engine & UI Architecture**

## Factor Analysis Diagnostic Pipeline — Mathematical Specification

### Z-Score Mapping Layer

$$
\hat{\mu}_j = \frac{1}{n}\sum_{i=1}^n x_{i,j}, \qquad
\hat{\sigma}_j = \max\!\left(\sqrt{\frac{1}{n-1}\sum_{i=1}^n(x_{i,j}-\hat{\mu}_j)^2},\; 10^{-15}\right)
$$

$$
Z_{i,j} = \frac{x_{i,j} - \hat{\mu}_j}{\hat{\sigma}_j}, \qquad
\mathbf{Z} \in \mathbb{R}^{n \times m}
$$

### Dimensionality Constraints

$$
n > m \quad \text{(snapshot count must exceed feature count)}
$$

$$
k < m \quad \text{(latent dimension must be strictly less than feature dimension)}
$$

### Information Decomposition — Fundamental Equation

$$
\mathbf{R} = \boldsymbol{\Lambda}_{\text{rot}}\boldsymbol{\Lambda}_{\text{rot}}^\top + \boldsymbol{\Psi}
$$

$$
h_j^2 = \sum_{r=1}^{k}\lambda_{j,r}^2, \qquad
\varphi_j^2 = 1 - h_j^2, \qquad
h_j^2 + \varphi_j^2 = 1 \quad \forall\, j
$$

### Varimax Rotation

$$
\boldsymbol{\Lambda}_{\text{rot}} = \boldsymbol{\Lambda}\,\mathbf{T}, \qquad
\mathbf{T}^\top\mathbf{T} = \mathbf{I}_{k\times k}
$$

$$
\mathbf{T}^* = \underset{\mathbf{T}:\,\mathbf{T}^\top\mathbf{T}=\mathbf{I}}{\arg\max}
\sum_{r=1}^{k}\left[
\frac{1}{m}\sum_{j=1}^{m}\lambda_{j,r}^{4}
- \left(\frac{1}{m}\sum_{j=1}^{m}\lambda_{j,r}^{2}\right)^{2}
\right]
$$

### Thomson's MMSE Factor Score Estimator

$$
\mathbf{R} = \mathbb{E}[\mathbf{Z}_i\mathbf{Z}_i^\top], \qquad
\mathbf{f}_i = \boldsymbol{\Lambda}_{\text{rot}}^\top\mathbf{R}^{-1}\mathbf{z}_i
$$

$$
\mathbf{F} = \mathbf{Z}\,\mathbf{R}^{-1}\,\boldsymbol{\Lambda}_{\text{rot}} \in \mathbb{R}^{n \times k}
$$

### Performance Telemetry

$$
\text{Average Communality} = \frac{1}{m}\sum_{j=1}^{m}h_j^2 \times 100\%
$$

$$
\text{Average Uniqueness} = \frac{1}{m}\sum_{j=1}^{m}\varphi_j^2 \times 100\%
$$

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import FactorAnalysis
from scipy.stats import ortho_group
import plotly.graph_objects as go
from plotly.subplots import make_subplots


# ── Varimax rotation ──────────────────────────────────────────────────────
def varimax(loadings: np.ndarray, max_iter: int = 1000, tol: float = 1e-6) -> np.ndarray:
    """
    Apply Varimax orthogonal rotation to a loading matrix.
    Returns rotated loading matrix (m x k).
    """
    m, k = loadings.shape
    T = np.eye(k)                          # rotation matrix initialised to identity

    for _ in range(max_iter):
        old_T = T.copy()
        L = loadings @ T                   # current rotated loadings (m x k)

        for j in range(k):
            lj = L[:, j]                   # j-th factor column
            # Varimax gradient step for column j
            u = lj ** 3 - (lj ** 2).mean() * lj
            g = loadings.T @ u             # (k,)
            T[:, j] = g / np.linalg.norm(g)

        # Re-orthogonalise via SVD  (keeps T exactly orthogonal)
        U, _, Vt = np.linalg.svd(T)
        T = U @ Vt

        if np.max(np.abs(T - old_T)) < tol:
            break

    return loadings @ T                    # (m x k) rotated matrix


# ── Main FA pipeline ──────────────────────────────────────────────────────
def factor_analysis_pipeline(df: pd.DataFrame, k: int):
    """
    End-to-end Factor Analysis diagnostic pipeline.

    Parameters
    ----------
    df : pd.DataFrame   Raw input dataframe (any mix of dtypes).
    k  : int            Number of latent factors to extract.
    """

    # ── Step 1: Z-Score Mapping Layer ─────────────────────────────────────
    numeric_df = df.select_dtypes(include=[np.number]).copy()
    sensor_names = list(numeric_df.columns)
    X = numeric_df.values.astype(float)          # (n, m)
    n, m = X.shape

    mu    = X.mean(axis=0)                        # (m,)
    sigma = X.std(axis=0, ddof=1)                 # unbiased std  (m,)
    sigma = np.where(sigma == 0, 1e-15, sigma)    # zero-variance guardrail
    Z     = (X - mu) / sigma                      # (n, m) standard normal space

    # ── Step 2: Dimensionality Constraint Layer ───────────────────────────
    if n <= m:
        raise ValueError(
            f"Snapshot count n={n} must be strictly greater than "
            f"feature count m={m}. Add more observations or reduce features."
        )
    if k >= m:
        raise ValueError(
            f"Latent dimension k={k} must be strictly less than "
            f"feature dimension m={m}. Choose k ∈ {{1, …, {m-1}}}."
        )

    # ── Step 3: Information Decomposition Layer ───────────────────────────
    fa = FactorAnalysis(n_components=k, svd_method='lapack', random_state=42)
    fa.fit(Z)

    Lambda_raw  = fa.components_.T              # (m, k)  raw loadings
    Lambda_rot  = varimax(Lambda_raw)           # (m, k)  Varimax-rotated loadings

    # Communality and Uniqueness
    h2   = (Lambda_rot ** 2).sum(axis=1)        # (m,)  h²_j = Σ_r λ²_{j,r}
    phi2 = 1.0 - h2                             # (m,)  φ²_j = 1 - h²_j
    phi2 = np.clip(phi2, 0.0, 1.0)             # numerical safety clip

    # ── Step 4: Thomson's MMSE Factor Score Reconstruction ───────────────
    # F = Z R⁻¹ Λ_rot   where R = (1/(n-1)) ZᵀZ  (empirical correlation)
    R     = np.cov(Z, rowvar=False, ddof=1)     # (m, m)
    R_inv = np.linalg.pinv(R)                   # pseudo-inverse for stability
    F     = Z @ R_inv @ Lambda_rot              # (n, k)  factor scores

    # ── Step 5: Performance Diagnostic Telemetry ─────────────────────────
    factor_var = F.var(axis=0, ddof=1)          # (k,)  unbiased variance per factor

    avg_communality = h2.mean()   * 100
    avg_uniqueness  = phi2.mean() * 100

    print("=" * 55)
    print("   FACTOR ANALYSIS DIAGNOSTIC TELEMETRY")
    print("=" * 55)
    print(f"  Sensors (m)              : {m}")
    print(f"  Snapshots (n)            : {n}")
    print(f"  Latent factors (k)       : {k}")
    print(f"  Average Communality      : {avg_communality:.2f} %")
    print(f"  Average Uniqueness       : {avg_uniqueness:.2f} %")
    print("-" * 55)
    for j, name in enumerate(sensor_names):
        print(f"  {name:<28}  h²={h2[j]:.4f}  φ²={phi2[j]:.4f}")
    print("-" * 55)
    for r in range(k):
        print(f"  Factor {r+1} empirical variance : {factor_var[r]:.4f}")
    print("=" * 55)

    # ── Step 6: 2×2 Diagnostic Canvas ────────────────────────────────────
    factor_labels  = [f"F{r+1}" for r in range(k)]
    h2_pct   = h2   * 100
    phi2_pct = phi2 * 100

    fig = make_subplots(
        rows=2, cols=2,
        horizontal_spacing=0.24,
        vertical_spacing=0.28,
        subplot_titles=(
            "Structural Loadings Matrix  |Λ|",
            "Variance Partitioning Profile",
            "Sensor Uniqueness Line Profile",
            "Latent Factor Variance Distribution",
        ),
    )

    # ── Subplot (1,1): Structural Loadings Heatmap ────────────────────────
    fig.add_trace(
        go.Heatmap(
            z=np.abs(Lambda_rot),
            x=factor_labels,
            y=sensor_names,
            colorscale="YlOrRd",
            colorbar=dict(
                title="| λ |",
                x=-0.15, xanchor="right",
                len=0.38, y=0.78, yanchor="middle",
            ),
            name="Loadings",
            showlegend=False,
        ),
        row=1, col=1,
    )
    fig.update_yaxes(autorange="reversed", row=1, col=1)

    # ── Subplot (1,2): Horizontal Stacked Bar — Variance Partition ────────
    fig.add_trace(
        go.Bar(
            y=sensor_names,
            x=h2_pct,
            orientation="h",
            name="Communality h²",
            marker_color="#1f77b4",
            legendgroup="comm",
        ),
        row=1, col=2,
    )
    fig.add_trace(
        go.Bar(
            y=sensor_names,
            x=phi2_pct,
            orientation="h",
            name="Uniqueness φ²",
            marker_color="#ff7f0e",
            legendgroup="uniq",
        ),
        row=1, col=2,
    )
    fig.update_xaxes(range=[0, 100], title_text="% Variance", row=1, col=2)
    fig.update_layout(barmode="stack")

    # ── Subplot (2,1): Sensor Uniqueness Line Profile ─────────────────────
    fig.add_trace(
        go.Scatter(
            x=sensor_names,
            y=phi2,
            mode="lines+markers",
            name="Uniqueness φ²",
            line=dict(color="#d62728", width=2, dash="dashdot"),
            marker=dict(symbol="x", size=8, color="#d62728"),
            showlegend=False,
        ),
        row=2, col=1,
    )
    fig.update_xaxes(tickangle=25, row=2, col=1)
    fig.update_yaxes(title_text="φ²", row=2, col=1)

    # ── Subplot (2,2): Latent Factor Variance Distribution ────────────────
    fig.add_trace(
        go.Bar(
            x=factor_labels,
            y=factor_var,
            name="Factor variance",
            marker=dict(
                color="#2ca02c",
                line=dict(color="black", width=0.5),
            ),
            showlegend=False,
        ),
        row=2, col=2,
    )
    fig.update_yaxes(title_text="Empirical variance", row=2, col=2)

    # ── Global layout ─────────────────────────────────────────────────────
    fig.update_layout(
        width=1250,
        height=750,
        template="plotly_white",
        legend=dict(
            orientation="h",
            x=0.5, xanchor="center",
            y=1.02, yanchor="bottom",
        ),
        margin=dict(t=150, b=60, l=140, r=80),
        title=dict(
            text="Factor Analysis Diagnostic Dashboard",
            x=0.5, xanchor="center", y=0.98,
        ),
    )

    fig.show()

    return {
        "Z"           : Z,
        "Lambda_raw"  : Lambda_raw,
        "Lambda_rot"  : Lambda_rot,
        "h2"          : h2,
        "phi2"        : phi2,
        "F"           : F,
        "factor_var"  : factor_var,
        "sensor_names": sensor_names,
    }


# ── Demo — synthetic dataset (replace with inspector.df) ─────────────────
np.random.seed(42)
n_demo, m_demo, k_demo = 500, 6, 2

true_Lambda = np.array([
    [0.9, 0.1], [0.85, 0.2], [0.8, 0.15],
    [0.1, 0.9], [0.15, 0.85], [0.2, 0.8],
])
true_Psi  = np.diag([0.15, 0.2, 0.18, 0.12, 0.22, 0.17])
true_cov  = true_Lambda @ true_Lambda.T + true_Psi

X_demo    = np.random.multivariate_normal(np.zeros(m_demo), true_cov, size=n_demo)
col_names = [f"Sensor_{j+1}" for j in range(m_demo)]
df_demo   = pd.DataFrame(X_demo, columns=col_names)

# ── Run pipeline ──────────────────────────────────────────────────────────
# Replace df_demo with inspector.df to use your real DataInspector dataset
results = factor_analysis_pipeline(df=df_demo, k=k_demo)
```

# **Subspace Diagnostics & Feature De-correlation**

## **Part A: Develop a Feature Analysis Dashboard**

## UI Layout 1 — PCA Optimization Suite

### Spectral Decomposition

$$
\mathbf{S} = \hat{\mathbf{P}}\,\hat{\boldsymbol{\Lambda}}\,\hat{\mathbf{P}}^\top, \qquad
\hat{\mathbf{P}}^\top\hat{\mathbf{P}} = \mathbf{I}_m, \qquad
\hat{\lambda}_1 \ge \hat{\lambda}_2 \ge \cdots \ge \hat{\lambda}_m \ge 0
$$

### Marginal and Cumulative Explained Variance

$$
\phi_j = \frac{\hat{\lambda}_j}{\sum_{l=1}^{m}\hat{\lambda}_l} \times 100\%, \qquad
\Phi(k) = \sum_{j=1}^{k}\phi_j, \qquad
\Psi(k) = 100 - \Phi(k)
$$

### Hotelling $T^2$ and $Q$ Statistics

$$
T^2_i(k) = \sum_{j=1}^{k}\frac{z_{i,j}^2}{\hat{\lambda}_j}, \qquad
Q_i(k) = \sum_{j=k+1}^{m} z_{i,j}^2
$$

$$
\overline{T^2}(k) = \frac{1}{n}\sum_{i=1}^{n}T^2_i(k), \qquad
\overline{Q}(k) = \frac{1}{n}\sum_{i=1}^{n}Q_i(k)
$$

---

## UI Layout 2 — FA Latent Subspace Suite

### Fundamental Equation of Factor Analysis

$$
\mathbf{R} = \boldsymbol{\Lambda}_{\text{rot}}\boldsymbol{\Lambda}_{\text{rot}}^\top + \boldsymbol{\Psi}, \qquad
\boldsymbol{\Psi} = \mathrm{diag}(\varphi_1^2,\dots,\varphi_m^2)
$$

### Communality and Uniqueness

$$
h_j^2 = \sum_{r=1}^{k}\lambda_{j,r}^2, \qquad
\varphi_j^2 = 1 - h_j^2, \qquad
h_j^2 + \varphi_j^2 = 1
$$

### Varimax Rotation

$$
\boldsymbol{\Lambda}_{\text{rot}} = \boldsymbol{\Lambda}\,\mathbf{T}, \qquad
\mathbf{T}^\top\mathbf{T} = \mathbf{I}_k
$$

### Thomson's MMSE Factor Score Estimator

$$
\mathbf{F} = \mathbf{Z}\,\mathbf{R}^{-1}\,\boldsymbol{\Lambda}_{\text{rot}} \in \mathbb{R}^{n \times k}
$$

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import FactorAnalysis
import plotly.graph_objects as go
from plotly.subplots import make_subplots


# ═══════════════════════════════════════════════════════════════════════════
#  UTILITIES
# ═══════════════════════════════════════════════════════════════════════════

def z_score(X: np.ndarray) -> np.ndarray:
    mu    = X.mean(axis=0)
    sigma = X.std(axis=0, ddof=1)
    sigma = np.where(sigma == 0, 1e-15, sigma)
    return (X - mu) / sigma


def varimax(L: np.ndarray, max_iter: int = 1000, tol: float = 1e-6) -> np.ndarray:
    """Varimax orthogonal rotation of loading matrix L (m x k)."""
    _, k = L.shape
    T = np.eye(k)
    for _ in range(max_iter):
        T_old = T.copy()
        Lr = L @ T
        for j in range(k):
            lj = Lr[:, j]
            g  = L.T @ (lj**3 - (lj**2).mean() * lj)
            T[:, j] = g / np.linalg.norm(g)
        U, _, Vt = np.linalg.svd(T)
        T = U @ Vt
        if np.max(np.abs(T - T_old)) < tol:
            break
    return L @ T


# ═══════════════════════════════════════════════════════════════════════════
#  LAYOUT 1 — PCA OPTIMIZATION SUITE  (2 × 3)
# ═══════════════════════════════════════════════════════════════════════════

def render_pca_suite(df: pd.DataFrame):

    # ── Data prep ──────────────────────────────────────────────────────────
    num_df       = df.select_dtypes(include=[np.number])
    sensor_names = list(num_df.columns)
    Z            = z_score(num_df.values.astype(float))
    n, m         = Z.shape

    S               = np.cov(Z, rowvar=False, ddof=1)
    eigenvalues, eigenvectors = np.linalg.eigh(S)
    idx             = np.argsort(eigenvalues)[::-1]
    lam             = eigenvalues[idx]
    P               = eigenvectors[:, idx]          # (m, m) sorted eigenvectors
    total_var       = lam.sum()

    marginal_pct    = lam / total_var * 100
    cumulative_pct  = np.cumsum(marginal_pct)
    residual_pct    = 100 - cumulative_pct
    k_axis          = np.arange(1, m + 1)

    # PC scores
    Zp = Z @ P                                      # (n, m)

    # Hotelling T² and Q for each k ∈ {1,…,m-1}
    T2_mean = []
    Q_mean  = []
    k_range = list(range(1, m))
    for k in k_range:
        T2 = (Zp[:, :k] ** 2 / lam[:k]).sum(axis=1)
        Q  = (Zp[:, k:]  ** 2).sum(axis=1)
        T2_mean.append(T2.mean())
        Q_mean.append(Q.mean())

    component_labels = [f"PC{j+1}" for j in range(m)]

    # ── Figure ─────────────────────────────────────────────────────────────
    fig = make_subplots(
        rows=2, cols=3,
        horizontal_spacing=0.18,
        vertical_spacing=0.30,
        subplot_titles=(
            "Feature Loadings  |P|",
            "Eigenvalue Spectrum  λ",
            "Marginal & Cumulative Variance %",
            "Residual Unexplained Variance %",
            "Mean Hotelling T²  vs  k",
            "Mean Q Statistic (SPE)  vs  k",
        ),
    )

    # Panel (1,1) — Loadings Heatmap
    fig.add_trace(
        go.Heatmap(
            z=np.abs(P),
            x=component_labels,
            y=sensor_names,
            colorscale="YlOrRd",
            colorbar=dict(
                title="|P|",
                x=-0.14, xanchor="right",
                len=0.38, y=0.78, yanchor="middle",
            ),
            showlegend=False,
        ),
        row=1, col=1,
    )
    fig.update_yaxes(autorange="reversed", row=1, col=1)

    # Panel (1,2) — Eigenvalue Bar
    fig.add_trace(
        go.Bar(
            x=component_labels,
            y=lam,
            name="Eigenvalue λ",
            marker=dict(color="#5b6abf", line=dict(color="black", width=0.5)),
            showlegend=False,
        ),
        row=1, col=2,
    )
    fig.update_yaxes(title_text="λ", row=1, col=2)

    # Panel (1,3) — Marginal bars + cumulative line
    fig.add_trace(
        go.Bar(
            x=component_labels,
            y=marginal_pct,
            name="Marginal variance %",
            marker=dict(color="#aec7e8", line=dict(color="black", width=0.5)),
            showlegend=True,
        ),
        row=1, col=3,
    )
    fig.add_trace(
        go.Scatter(
            x=component_labels,
            y=cumulative_pct,
            mode="lines+markers",
            name="Cumulative variance %",
            line=dict(color="#d62728", width=2, dash="dash"),
            marker=dict(size=6),
        ),
        row=1, col=3,
    )
    fig.update_yaxes(title_text="%", range=[0, 105], row=1, col=3)

    # Panel (2,1) — Residual bar
    fig.add_trace(
        go.Bar(
            x=component_labels,
            y=residual_pct,
            name="Residual variance %",
            marker=dict(color="#ffbb78", line=dict(color="black", width=0.5)),
            showlegend=False,
        ),
        row=2, col=1,
    )
    fig.update_yaxes(title_text="Unexplained %", row=2, col=1)

    # Panel (2,2) — Mean T² vs k
    fig.add_trace(
        go.Scatter(
            x=k_range,
            y=T2_mean,
            mode="lines+markers",
            name="Mean T²",
            line=dict(color="#1f77b4", width=2),
            marker=dict(size=7, symbol="circle"),
            showlegend=False,
        ),
        row=2, col=2,
    )
    fig.update_xaxes(title_text="Subspace size k", row=2, col=2)
    fig.update_yaxes(title_text="E[T²]",           row=2, col=2)

    # Panel (2,3) — Mean Q vs k
    fig.add_trace(
        go.Scatter(
            x=k_range,
            y=Q_mean,
            mode="lines+markers",
            name="Mean Q (SPE)",
            line=dict(color="#2ca02c", width=2),
            marker=dict(size=7, symbol="diamond"),
            showlegend=False,
        ),
        row=2, col=3,
    )
    fig.update_xaxes(title_text="Truncation cutoff k", row=2, col=3)
    fig.update_yaxes(title_text="E[Q]",               row=2, col=3)

    fig.update_layout(
        width=1350, height=780,
        template="plotly_white",
        title=dict(text="PCA Optimization Suite", x=0.5, xanchor="center"),
        legend=dict(orientation="h", x=0.5, xanchor="center", y=1.04, yanchor="bottom"),
        margin=dict(t=160, b=60, l=150, r=80),
    )
    fig.show()


# ═══════════════════════════════════════════════════════════════════════════
#  LAYOUT 2 — FA LATENT SUBSPACE SUITE  (2 × 2)
# ═══════════════════════════════════════════════════════════════════════════

def render_fa_suite(df: pd.DataFrame, k: int):

    # ── Data prep ──────────────────────────────────────────────────────────
    num_df       = df.select_dtypes(include=[np.number])
    sensor_names = list(num_df.columns)
    Z            = z_score(num_df.values.astype(float))
    n, m         = Z.shape

    if n <= m:
        raise ValueError(f"n={n} must be > m={m}.")
    if k >= m:
        raise ValueError(f"k={k} must be < m={m}.")

    fa = FactorAnalysis(n_components=k, svd_method='lapack', random_state=42)
    fa.fit(Z)
    Lambda_rot  = varimax(fa.components_.T)         # (m, k)

    h2   = (Lambda_rot ** 2).sum(axis=1)
    phi2 = np.clip(1.0 - h2, 0.0, 1.0)

    R        = np.cov(Z, rowvar=False, ddof=1)
    F_scores = Z @ np.linalg.pinv(R) @ Lambda_rot   # (n, k) Thomson MMSE
    fvar     = F_scores.var(axis=0, ddof=1)          # (k,)

    factor_labels = [f"F{r+1}" for r in range(k)]

    # ── Console telemetry ──────────────────────────────────────────────────
    print("=" * 52)
    print("   FA LATENT SUBSPACE TELEMETRY")
    print("=" * 52)
    print(f"  Sensors (m)         : {m}")
    print(f"  Snapshots (n)       : {n}")
    print(f"  Latent factors (k)  : {k}")
    print(f"  Avg Communality     : {h2.mean()*100:.2f} %")
    print(f"  Avg Uniqueness      : {phi2.mean()*100:.2f} %")
    print("-" * 52)
    for j, name in enumerate(sensor_names):
        print(f"  {name:<26}  h²={h2[j]:.4f}  φ²={phi2[j]:.4f}")
    print("=" * 52)

    # ── Figure ─────────────────────────────────────────────────────────────
    fig = make_subplots(
        rows=2, cols=2,
        horizontal_spacing=0.24,
        vertical_spacing=0.28,
        subplot_titles=(
            "Structural Loadings  |Λ|",
            "Variance Allocation per Sensor",
            "Sensor Uniqueness Noise Floor  φ²",
            "Latent Factor Empirical Variance",
        ),
    )

    # Panel (1,1) — Loadings Heatmap
    fig.add_trace(
        go.Heatmap(
            z=np.abs(Lambda_rot),
            x=factor_labels,
            y=sensor_names,
            colorscale="YlOrRd",
            colorbar=dict(
                title="|λ|",
                x=-0.15, xanchor="right",
                len=0.38, y=0.78, yanchor="middle",
            ),
            showlegend=False,
        ),
        row=1, col=1,
    )
    fig.update_yaxes(autorange="reversed", row=1, col=1)

    # Panel (1,2) — Stacked horizontal bar
    fig.add_trace(
        go.Bar(
            y=sensor_names,
            x=h2 * 100,
            orientation="h",
            name="Communality h²",
            marker_color="#1f77b4",
        ),
        row=1, col=2,
    )
    fig.add_trace(
        go.Bar(
            y=sensor_names,
            x=phi2 * 100,
            orientation="h",
            name="Uniqueness φ²",
            marker_color="#ff7f0e",
        ),
        row=1, col=2,
    )
    fig.update_xaxes(range=[0, 100], title_text="% Variance", row=1, col=2)
    fig.update_layout(barmode="stack")

    # Panel (2,1) — Uniqueness line
    fig.add_trace(
        go.Scatter(
            x=sensor_names,
            y=phi2,
            mode="lines+markers",
            name="Uniqueness φ²",
            line=dict(color="#d62728", width=2, dash="dashdot"),
            marker=dict(symbol="x", size=8, color="#d62728"),
            showlegend=False,
        ),
        row=2, col=1,
    )
    fig.update_xaxes(tickangle=25,    row=2, col=1)
    fig.update_yaxes(title_text="φ²", row=2, col=1)

    # Panel (2,2) — Factor variance bar
    fig.add_trace(
        go.Bar(
            x=factor_labels,
            y=fvar,
            name="Factor variance",
            marker=dict(color="#2ca02c", line=dict(color="black", width=0.5)),
            showlegend=False,
        ),
        row=2, col=2,
    )
    fig.update_yaxes(title_text="Empirical variance", row=2, col=2)

    fig.update_layout(
        width=1250, height=750,
        template="plotly_white",
        title=dict(text="FA Latent Subspace Suite", x=0.5, xanchor="center"),
        legend=dict(orientation="h", x=0.5, xanchor="center", y=1.02, yanchor="bottom"),
        margin=dict(t=150, b=60, l=140, r=80),
    )
    fig.show()


# ═══════════════════════════════════════════════════════════════════════════
#  DEMO — replace df_demo with inspector.df
# ═══════════════════════════════════════════════════════════════════════════

np.random.seed(42)
n_demo, m_demo, k_demo = 600, 7, 2

true_L   = np.random.randn(m_demo, k_demo) * 0.7
true_Psi = np.diag(np.random.uniform(0.1, 0.4, m_demo))
true_cov = true_L @ true_L.T + true_Psi

X_demo   = np.random.multivariate_normal(np.zeros(m_demo), true_cov, size=n_demo)
cols     = [f"Sensor_{j+1}" for j in range(m_demo)]
df_demo  = pd.DataFrame(X_demo, columns=cols)

render_pca_suite(df=df_demo)
render_fa_suite(df=df_demo, k=k_demo)
```

```python
import numpy as np
import pandas as pd
from typing import Optional, Sequence, Dict, Any
from sklearn.decomposition import FactorAnalysis
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# =====================================================================
# COMBINED SUBSPACE ENGINES (PCA & FA PIPELINES)
# =====================================================================

class DualSubspaceDiagnosticsEngine:
    def __init__(self, df: pd.DataFrame):
        self.df = df

    def compute_empirical_pca(self, columns: Optional[Sequence[str]] = None, show_plot: bool = True) -> Dict[str, Any]:
        if self.df is None:
            raise ValueError("Error: No data loaded.")

        if columns is None:
            target_cols = self.df.select_dtypes(include=[np.number]).columns.tolist()
            if 'count' in target_cols: target_cols.remove('count')
            target_cols = list(target_cols)
        else:
            target_cols = list(columns)

        X = self.df[target_cols].copy().dropna().values
        n, m = X.shape

        if n <= m:
            raise ValueError("Sample size n must be larger than feature dimensions m to isolate principal components.")

        mu_hat      = np.mean(X, axis=0)
        X_centered  = X - mu_hat
        S_matrix    = np.cov(X, rowvar=False, ddof=1)

        eigenvalues, eigenvectors = np.linalg.eigh(S_matrix)
        idx         = np.argsort(eigenvalues)[::-1]
        lambda_hat  = eigenvalues[idx]
        P_hat       = eigenvectors[:, idx]
        lambda_hat  = np.clip(lambda_hat, a_min=1e-15, a_max=None)

        total_variance          = np.sum(lambda_hat)
        if total_variance == 0:
            raise ValueError("Total variance is zero.")

        explained_variance_ratio   = lambda_hat / total_variance
        cumulative_variance_ratio  = np.cumsum(explained_variance_ratio)
        unexplained_variance_ratio = 1.0 - cumulative_variance_ratio

        Z_scores = np.dot(X_centered, P_hat)
        S_Z      = np.cov(Z_scores, rowvar=False, ddof=1)

        k_range     = np.arange(1, m)
        mean_T2_vs_k = []
        mean_Q_vs_k  = []
        T2_matrix    = np.zeros((n, len(k_range)))
        Q_matrix     = np.zeros((n, len(k_range)))

        for idx_k, k in enumerate(k_range):
            Z_k        = Z_scores[:, :k]
            lambda_k   = lambda_hat[:k]
            T2_samples = np.sum((Z_k ** 2) / lambda_k, axis=1)
            T2_matrix[:, idx_k] = T2_samples
            mean_T2_vs_k.append(np.mean(T2_samples))

            Z_residual = Z_scores[:, k:]
            Q_samples  = np.sum(Z_residual ** 2, axis=1)
            Q_matrix[:, idx_k] = Q_samples
            mean_Q_vs_k.append(np.mean(Q_samples))

        print(f"\n--- Operationalizing Geometric De-correlation Layer (PCA) ---")
        print(f"Decomposing structural space of {m} features using {n} samples.")
        print(f"Total System Variance (Trace[S]): {total_variance:.4f}")

        if show_plot:
            pc_labels     = [f"PC {i+1}" for i in range(m)]
            k_labels      = [f"k={k}"    for k in k_range]
            sensor_labels = target_cols

            fig = make_subplots(
                rows=2, cols=3,
                horizontal_spacing=0.18,
                vertical_spacing=0.28,
                subplot_titles=(
                    "Feature Loading Matrix |P_hat|",
                    "Component Values (Eigenvalues λ)",
                    "Information Profile (Explained Var.)",
                    "Residual Space (Unexplained Var.)",
                    "Mean Hotelling's T² vs Subspace Size k",
                    "Mean Q Statistic (SPE) vs Subspace Size k",
                ),
            )

            # ── (1,1) Loading Heatmap — Viridis ──────────────────────────
            fig.add_trace(
                go.Heatmap(
                    z=np.abs(P_hat),
                    x=pc_labels,
                    y=sensor_labels,
                    colorscale="Viridis",
                    colorbar=dict(
                        title="Loading Weight",
                        x=-0.12, xanchor="right",
                        len=0.38, y=0.78, yanchor="middle",
                        titleside="top",
                        tickfont=dict(size=10),
                        titlefont=dict(size=11),
                    ),
                    showscale=True,
                    showlegend=False,
                ),
                row=1, col=1,
            )
            fig.update_xaxes(title_text="Principal Axes", row=1, col=1)
            fig.update_yaxes(autorange="reversed",        row=1, col=1)

            # ── (1,2) Eigenvalue Bar — deep teal ─────────────────────────
            fig.add_trace(
                go.Bar(
                    x=pc_labels, y=lambda_hat,
                    name="Eigenvalue (λ_j)",
                    marker=dict(
                        color=lambda_hat,
                        colorscale="Teal",
                        showscale=False,
                        line=dict(color="#1a3a3a", width=0.8),
                    ),
                    legendgroup="eigen",
                ),
                row=1, col=2,
            )
            fig.update_yaxes(title_text="Variance Magnitude", row=1, col=2)
            fig.update_xaxes(title_text="Principal Axes",     row=1, col=2)

            # ── (1,3) Marginal bars + cumulative line ─────────────────────
            fig.add_trace(
                go.Bar(
                    x=pc_labels, y=explained_variance_ratio * 100,
                    name="Marginal Explained",
                    marker=dict(color="#7b5ea7", opacity=0.82,
                                line=dict(color="#3d2b5e", width=0.7)),
                    legendgroup="expl",
                ),
                row=1, col=3,
            )
            fig.add_trace(
                go.Scatter(
                    x=pc_labels, y=cumulative_variance_ratio * 100,
                    mode="lines+markers",
                    name="Cumulative Captured",
                    line=dict(color="#f0a500", width=2.5, dash="dash"),
                    marker=dict(size=7, symbol="circle",
                                color="#f0a500",
                                line=dict(color="#7a5000", width=1)),
                    legendgroup="expl",
                ),
                row=1, col=3,
            )
            fig.update_yaxes(title_text="Captured Structure (%)", range=[-2, 105], row=1, col=3)
            fig.update_xaxes(title_text="Principal Axes",          row=1, col=3)

            # ── (2,1) Residual bar — warm coral ───────────────────────────
            fig.add_trace(
                go.Bar(
                    x=pc_labels, y=unexplained_variance_ratio * 100,
                    name="Remaining Noise",
                    marker=dict(
                        color=unexplained_variance_ratio * 100,
                        colorscale="Reds",
                        showscale=False,
                        line=dict(color="#5a1a00", width=0.7),
                    ),
                    legendgroup="noise",
                ),
                row=2, col=1,
            )
            fig.update_yaxes(title_text="Excluded Info (%)", range=[-2, 105], row=2, col=1)
            fig.update_xaxes(title_text="Principal Axes",    row=2, col=1)

            # ── (2,2) Hotelling T² — indigo ───────────────────────────────
            fig.add_trace(
                go.Scatter(
                    x=k_labels, y=mean_T2_vs_k,
                    mode="lines+markers",
                    name="Mean T²",
                    line=dict(color="#3d5a80", width=2.5),
                    marker=dict(size=8, symbol="diamond",
                                color="#3d5a80",
                                line=dict(color="#0d1b2a", width=1)),
                    legendgroup="t2",
                ),
                row=2, col=2,
            )
            fig.update_yaxes(title_text="Average T² Metric",    row=2, col=2)
            fig.update_xaxes(title_text="Truncation Cutoff (k)", row=2, col=2)

            # ── (2,3) Q / SPE — magenta-rose ─────────────────────────────
            fig.add_trace(
                go.Scatter(
                    x=k_labels, y=mean_Q_vs_k,
                    mode="lines+markers",
                    name="Mean Q (SPE)",
                    line=dict(color="#c9184a", width=2.5),
                    marker=dict(size=8, symbol="square",
                                color="#c9184a",
                                line=dict(color="#590d22", width=1)),
                    legendgroup="q_stat",
                ),
                row=2, col=3,
            )
            fig.update_yaxes(title_text="Average Residual Energy",  row=2, col=3)
            fig.update_xaxes(title_text="Truncation Cutoff (k)",    row=2, col=3)

            # ── Global layout ─────────────────────────────────────────────
            fig.update_layout(
                title=dict(
                    text="Principal Component Analysis (PCA) Optimization & Feature Loading Dashboard",
                    x=0.5, y=0.97, xanchor="center", yanchor="top",
                    font=dict(size=16, color="#1a1a2e"),
                ),
                template="plotly_white",
                showlegend=True,
                legend=dict(
                    orientation="h",
                    yanchor="bottom", y=1.02,
                    xanchor="center", x=0.5,
                    font=dict(size=11),
                    bgcolor="rgba(255,255,255,0.85)",
                    bordercolor="#cccccc", borderwidth=0.8,
                ),
                margin=dict(t=150, b=60, l=140, r=80),
                height=750, width=1250,
                plot_bgcolor="#f9f9fb",
                paper_bgcolor="#ffffff",
            )
            fig.update_xaxes(showgrid=True, gridcolor="#e8e8ee", gridwidth=0.5)
            fig.update_yaxes(showgrid=True, gridcolor="#e8e8ee", gridwidth=0.5)
            fig.show()

        return {
            "mean_vector": mu_hat, "covariance_matrix_S": S_matrix,
            "eigenvalues_lambda": lambda_hat, "eigenvectors_P": P_hat,
            "explained_variance_ratio": explained_variance_ratio,
            "cumulative_variance_ratio": cumulative_variance_ratio,
            "unexplained_variance_ratio": unexplained_variance_ratio,
            "transformed_scores_Z": Z_scores,
            "score_covariance_diagonal": np.diag(S_Z),
            "features": target_cols, "k_values": k_range,
            "T2_matrix_vs_k": T2_matrix, "Q_matrix_vs_k": Q_matrix,
            "mean_T2_profile": np.array(mean_T2_vs_k),
            "mean_Q_profile": np.array(mean_Q_vs_k),
        }

    def compute_empirical_fa(self, k: int, columns: Optional[Sequence[str]] = None, show_plot: bool = True) -> Dict[str, Any]:
        if self.df is None:
            raise ValueError("Error: No data loaded.")

        if columns is None:
            target_cols = self.df.select_dtypes(include=[np.number]).columns.tolist()
            if 'count' in target_cols: target_cols.remove('count')
            target_cols = list(target_cols)
        else:
            target_cols = list(columns)

        X = self.df[target_cols].copy().dropna().values
        n, m = X.shape

        mu_hat  = np.mean(X, axis=0)
        std_hat = np.std(X, axis=0, ddof=1)
        std_hat[std_hat == 0] = 1e-15
        Z        = (X - mu_hat) / std_hat
        R_matrix = np.corrcoef(X, rowvar=False)

        fa = FactorAnalysis(n_components=k, rotation='varimax', random_state=42)
        fa.fit(Z)
        lambda_matrix = fa.components_.T
        psi_diagonal  = fa.noise_variance_

        communality = np.sum(lambda_matrix ** 2, axis=1)
        uniqueness  = psi_diagonal

        F_scores = fa.transform(Z)

        print(f"\n--- Operationalizing Latent Subspace Layer (Factor Analysis) ---")
        print(f"Decomposing structural space of {m} sensors into {k} hidden physical factors.")
        print(f"Average System Communality (Shared Subspace Energy): {np.mean(communality)*100:.2f}%")
        print(f"Average System Uniqueness (Localized Noise Floor):   {np.mean(uniqueness)*100:.2f}%")

        if show_plot:
            sensor_labels = target_cols
            factor_labels = [f"Factor {j+1}" for j in range(k)]

            fig = make_subplots(
                rows=2, cols=2,
                horizontal_spacing=0.24,
                vertical_spacing=0.28,
                subplot_titles=(
                    "Structural Loadings Matrix |λ_(j,r)|",
                    "Variance Partitioning (Communality vs Uniqueness)",
                    "Sensor Uniqueness Noise Floor (φ²)",
                    "Latent Factor Scores Empirical Variance",
                ),
            )

            # ── (1,1) Loadings Heatmap — Plasma ──────────────────────────
            fig.add_trace(
                go.Heatmap(
                    z=np.abs(lambda_matrix),
                    x=factor_labels,
                    y=sensor_labels,
                    colorscale="Plasma",
                    colorbar=dict(
                        title="Sensitivity",
                        x=-0.10, xanchor="right",
                        len=0.38, y=0.78, yanchor="middle",
                        tickfont=dict(size=10),
                        titlefont=dict(size=11),
                    ),
                    showscale=True,
                    showlegend=False,
                ),
                row=1, col=1,
            )
            fig.update_xaxes(title_text="Latent Structures", row=1, col=1)
            fig.update_yaxes(autorange="reversed",           row=1, col=1)

            # ── (1,2) Stacked horizontal bar ──────────────────────────────
            fig.add_trace(
                go.Bar(
                    y=sensor_labels, x=communality * 100,
                    name="Communality (h² — Shared Structure)",
                    orientation="h",
                    marker=dict(color="#0077b6",
                                line=dict(color="#03045e", width=0.6)),
                    legendgroup="communality",
                ),
                row=1, col=2,
            )
            fig.add_trace(
                go.Bar(
                    y=sensor_labels, x=uniqueness * 100,
                    name="Uniqueness (φ² — Channel Noise)",
                    orientation="h",
                    marker=dict(color="#f77f00",
                                line=dict(color="#7a3900", width=0.6)),
                    legendgroup="uniqueness_bar",
                ),
                row=1, col=2,
            )
            fig.update_layout(barmode="stack")
            fig.update_xaxes(title_text="Variance Allocation (%)", range=[0, 100], row=1, col=2)

            # ── (2,1) Uniqueness line — crimson dot-dash ──────────────────
            fig.add_trace(
                go.Scatter(
                    x=sensor_labels, y=uniqueness,
                    mode="lines+markers",
                    name="Uniqueness Profile (φ²)",
                    line=dict(color="#e63946", width=2.2, dash="dot"),
                    marker=dict(size=9, symbol="x", color="#e63946",
                                line=dict(color="#7d0000", width=1.5)),
                    legendgroup="uniqueness_line",
                ),
                row=2, col=1,
            )
            fig.update_yaxes(range=[-0.05, 1.05], title_text="φ²",  row=2, col=1)
            fig.update_xaxes(title_text="Monitored Channels", tickangle=25, row=2, col=1)

            # ── (2,2) Factor variance bar — forest green ──────────────────
            factor_variances = np.var(F_scores, axis=0, ddof=1)
            fig.add_trace(
                go.Bar(
                    x=factor_labels, y=factor_variances,
                    name="Factor Empirical Variance",
                    marker=dict(
                        color=factor_variances,
                        colorscale="Greens",
                        showscale=False,
                        line=dict(color="#1b4332", width=0.8),
                    ),
                    legendgroup="variance",
                ),
                row=2, col=2,
            )
            fig.update_yaxes(title_text="Variance Level",  row=2, col=2)
            fig.update_xaxes(title_text="Latent Vectors",  row=2, col=2)

            # ── Global layout ─────────────────────────────────────────────
            fig.update_layout(
                title=dict(
                    text="Factor Analysis (FA) Latent Subspace Diagnostics Dashboard",
                    x=0.5, y=0.97, xanchor="center", yanchor="top",
                    font=dict(size=16, color="#1a1a2e"),
                ),
                template="plotly_white",
                showlegend=True,
                legend=dict(
                    orientation="h",
                    yanchor="bottom", y=1.02,
                    xanchor="center", x=0.5,
                    traceorder="normal",
                    itemwidth=40, itemsizing="constant",
                    font=dict(size=11),
                    bgcolor="rgba(255,255,255,0.85)",
                    bordercolor="#cccccc", borderwidth=0.8,
                ),
                margin=dict(t=100, b=60, l=80, r=80),
                height=750, width=1100,
                plot_bgcolor="#f9f9fb",
                paper_bgcolor="#ffffff",
            )
            fig.update_xaxes(showgrid=True, gridcolor="#e8e8ee", gridwidth=0.5)
            fig.update_yaxes(showgrid=True, gridcolor="#e8e8ee", gridwidth=0.5)
            fig.show()

        return {
            "mean_vector_mu": mu_hat, "std_vector_D": std_hat,
            "correlation_matrix_R": R_matrix,
            "factor_loadings_lambda": lambda_matrix,
            "uniqueness_psi": uniqueness, "communality_h2": communality,
            "latent_factor_scores_F": F_scores, "sensors": target_cols,
        }


# =====================================================================
# DATA GENERATOR & PIPELINE EXECUTION
# =====================================================================
if __name__ == "__main__":
    np.random.seed(42)
    n_samples = 2500

    f1 = np.random.normal(0, 1, n_samples)
    f2 = np.random.normal(0, 1, n_samples)

    s1 = 0.85 * f1 + 0.10 * f2 + np.random.normal(0, 0.30, n_samples)
    s2 = 0.80 * f1 + 0.15 * f2 + np.random.normal(0, 0.35, n_samples)
    s3 = 0.12 * f1 + 0.90 * f2 + np.random.normal(0, 0.25, n_samples)
    s4 = 0.02 * f1 + 0.05 * f2 + np.random.normal(0, 1.40, n_samples)

    df_asset = pd.DataFrame(
        data=np.vstack([s1, s2, s3, s4]).T,
        columns=['Sensor_1', 'Sensor_2', 'Sensor_3', 'Sensor_4'],
    )

    engine = DualSubspaceDiagnosticsEngine(df_asset)
    pca_metrics = engine.compute_empirical_pca(show_plot=True)
    fa_metrics  = engine.compute_empirical_fa(k=2, show_plot=True)
```

### Question 1: The Total Variance Illusion

## Diagnostic Interpretation — Sensor 4 Fault Analysis

### Part 1 — Physical Nature of Sensor 4

From the Factor Analysis uniqueness decomposition, the $j$-th diagonal entry of the
fundamental equation $\mathbf{R} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^\top + \boldsymbol{\Psi}$
satisfies:

$$
h_j^2 + \varphi_j^2 = 1, \qquad
\varphi_j^2 = 1 - \sum_{r=1}^{k}\lambda_{j,r}^2
$$

For \texttt{Sensor\_4}, the empirical fit returns $\varphi_4^2 \to 1.0$, which implies:

$$
h_4^2 = \sum_{r=1}^{k}\lambda_{4,r}^2 \approx 0
\qquad \Longrightarrow \qquad
\boldsymbol{\lambda}_4 \approx \mathbf{0}_k
$$

Geometrically, the loading row vector $\boldsymbol{\lambda}_4^\top \in \mathbb{R}^k$ has collapsed
to the zero vector, meaning \texttt{Sensor\_4} projects zero energy onto the shared
structural subspace $\operatorname{col}(\boldsymbol{\Lambda})$. Its entire unit variance budget
is absorbed by the isolated noise term:

$$
\mathbb{E}[Z_{i,4}^2] = \underbrace{h_4^2}_{\approx\; 0} + \underbrace{\varphi_4^2}_{\approx\; 1.0} = 1
$$

The generative model $Z_{i,4} = \boldsymbol{\lambda}_4^\top \mathbf{F}_i + \epsilon_{i,4}$
therefore degenerates to $Z_{i,4} \approx \epsilon_{i,4}$, confirming that
\texttt{Sensor\_4} is statistically orthogonal to every other channel in the array.
Its data profile consists entirely of isolated white noise or severe electrical
fault energy, completely decoupled from the true physical processes $(f_1, f_2)$
driving the plant asset.

---

### Part 2 — PCA Loading Contamination

The PCA spectral decomposition solves the constrained variance-maximization problem:

$$
\hat{\mathbf{p}}_1 = \underset{\|\mathbf{v}\|=1}{\arg\max}\; \mathbf{v}^\top \mathbf{S}\, \mathbf{v}
$$

Because \texttt{Sensor\_4} is generated as:

$$
s_4 = 0.02\, f_1 + 0.05\, f_2 + \varepsilon_4, \qquad \varepsilon_4 \sim \mathcal{N}(0,\, 1.40^2)
$$

its marginal sample variance dominates the empirical covariance spectrum:

$$
\hat{\sigma}_4^2 \approx (0.02)^2 \cdot 1 + (0.05)^2 \cdot 1 + 1.40^2 \approx 1.961 \gg \hat{\sigma}_j^2 \;\text{ for } j \in \{1,2,3\}
$$

The unbiased sample covariance matrix $\mathbf{S}$ therefore concentrates its largest
eigenvalue along the direction of \texttt{Sensor\_4}'s variance axis:

$$
\hat{\lambda}_1 \approx 2.0, \qquad
\hat{\mathbf{p}}_1 \approx \mathbf{e}_4 = [0,\; 0,\; 0,\; 1]^\top
$$

This manifests in the Feature Loading Matrix $|\hat{\mathbf{P}}|$ as a deep-red block
at position $(4, \text{PC\,1})$ with loading magnitude $|\hat{P}_{4,1}| \approx 1.0$,
while all remaining sensors carry near-zero projections onto the primary axis:

$$
|\hat{P}_{j,1}| \approx 0 \quad \forall\; j \in \{1, 2, 3\}
$$

The cumulative explained variance ratio $\Phi(k)$ is therefore corrupted at $k=1$:

$$
\Phi(1) = \frac{\hat{\lambda}_1}{\sum_{l=1}^{m}\hat{\lambda}_l}
\approx \frac{2.0}{2.0 + \hat{\lambda}_2 + \hat{\lambda}_3 + \hat{\lambda}_4}
$$

This inflation suppresses $\hat{\lambda}_2$ and $\hat{\lambda}_3$, which genuinely
encode the shared structural dynamics $(f_1, f_2)$, forcing them into subordinate
principal axes where they become invisible to a standard single-component monitor.

---

### Part 3 — Engineering Risk and Operational Hazard

Let the $k$-component Hotelling $T^2$ control statistic be defined as:

$$
T_i^2(k) = \sum_{j=1}^{k} \frac{z_{i,j}^2}{\hat{\lambda}_j}
$$

Under nominal conditions with a healthy sensor array, $T_i^2(1)$ tracks the dominant
structural mode $f_1$. After \texttt{Sensor\_4} injects fault energy, the primary
axis realigns to track pure electrical noise instead:

$$
T_i^2(1)\big|_{\text{contaminated}}
= \frac{z_{i,\text{PC1}}^2}{\hat{\lambda}_1}
\approx \frac{(\hat{\mathbf{p}}_1^\top \tilde{\mathbf{x}}_i)^2}{2.0}
\approx \frac{\tilde{x}_{i,4}^2}{2.0}
$$

The control chart now monitors the variance of \texttt{Sensor\_4} exclusively.
A true systemic structural failure propagating through $(f_1, f_2)$ into
\texttt{Sensor\_1}, \texttt{Sensor\_2}, and \texttt{Sensor\_3} produces no
displacement along $\hat{\mathbf{p}}_1$, and therefore generates no alarm signal:

$$
\Delta T^2(1)\big|_{\text{true fault}} \approx 0
\qquad \text{(false negative — missed detection)}
$$

Simultaneously, the $Q$ statistic at $k=1$:

$$
Q_i(1) = \sum_{j=2}^{m} z_{i,j}^2
$$

accumulates the true structural dynamics into the residual subspace where they are
treated as noise rather than signal. The formal consequence is that the
**size** of the monitoring test (Type I error rate $\alpha$) and the
**power** of detection (Type II complement $1 - \beta$) are both degraded:

$$
\sup_{\theta \in \Theta_0} P_\theta\!\left(T^2(1) > c_\alpha\right) = \alpha
\quad \text{is met, but} \quad
P_{\theta \in \Theta_1}\!\left(T^2(1) > c_\alpha\right) \approx \alpha
$$

meaning the test has power indistinguishable from its size — it provides no
discriminative capability against true structural anomalies. A single blown or
short-circuited instrument therefore hijacks the most critical monitoring axis,
inflating $\hat{\lambda}_1 \approx 2.0$, masking the true common structural
dynamics of the asset, and rendering the primary Hotelling $T^2$ control chart
operationally useless for catching systemic failures.

The Factor Analysis framework resolves this precisely because its likelihood
objective explicitly separates shared covariance structure from isolated channel
noise, preventing any single high-variance sensor from contaminating the shared
loading subspace $\boldsymbol{\Lambda}\boldsymbol{\Lambda}^\top$.

###  Question 2: Decoupling Structural Loading via Rotation

## Part 1: Mathematical Jumble vs. Simple Structure

Because PCA is bound to strict, orthogonal variance maximization, consider the behavior of the remaining signals on $PC_2$ and $PC_4$.

$PC_2$ and $PC_4$ exhibit high loading values across both $Sensor_1$ and $Sensor_2$, while also bleeding into $Sensor_3$.

This cross-loading represents a mathematically forced mixture of signals.

Factor Analysis applies a **Varimax rotation** to eliminate this effect.

By maximizing loading variance within columns, it separates the structure more cleanly:
- Sensor_1 and Sensor_2 group into Factor 1
- Sensor_3 shifts into Factor 2

---

## Part 2: Operational Interpretability

The PCA loading matrix forces a complex interpretation.

If a diagnostic alarm is triggered on $PC_2$, an engineer must interpret a blended contribution from Sensor_1, Sensor_2, and Sensor_3.

With Factor Analysis:
- Factor 1 → redundant physical process in Sensor_1 & Sensor_2  
- Factor 2 → independent process in Sensor_3  

This makes root-cause analysis much more direct.

### Question 3: Determining Subspace Truncation ($k$)

Part 1 (Curve Observations): Moving from $k=1$ to $k=2$, the Mean Q Statistic (SPE) plunges from 2.4 down to 0.9. When transitioning from $k=2$ to $k=3$, the curve flattens out into an explicit "elbow", decreasing at a much slower, almost linear rate down toward 0.1.
Part 2 (The True Intrinsic Dimension): Cross-referencing the Q curve with the new loading matrix makes the selection clear. PC1 tracks the isolated noise of Sensor_4. PC2 captures the true, shared variance of the working sensors (Sensor_1, Sensor_2, Sensor_3). The sharp drop in the Q statistic at $k=2$ tells us that once these two distinct behaviors are accounted for, the structured information is fully exhausted. Thus, $k=2$ represents the true hidden physical dimensionality of the asset.
Part 3 (The Danger of Over-fitting to k=3): Look at PC3 in the loading matrix—it maps exclusively to the leftover residual variance of Sensor_3. Setting your truncation limit to $k=3$ pulls this residual space into your "clean" model subspace. This overfits the system, causing Hotelling's T² control limits to trigger frequent false alarms due to minor, normal baseline variations.

### Question 4: Operational Trade-offs in System Health Monitoring

The PCA Strategy: Tracking the main principal subspace via $T^2$ alongside the residual space via the Q statistic.
Vulnerability: This approach is highly vulnerable to sensor degradation. As shown in the PCA loading matrix, if a single channel (Sensor_4) fails high, it can structurally shift the orientation of the principal axes, dragging the calculations for the remaining healthy channels along with it.
The FA Strategy: Directly monitoring the rotated latent factor scores.
Robustness: This approach is significantly more robust. Because Factor Analysis explicitly separates localized noise via the diagonal uniqueness vector ($\Psi$), any single-channel sensor failure is absorbed by that channel’s uniqueness term ($\psi_4^2 \rightarrow 1$). The underlying latent factors (Factor1 and Factor2) remain clean, stable, and uncorrupted, allowing the rest of the health monitoring network to function normally.
