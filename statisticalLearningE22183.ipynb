\# Assignment: Probability on a Disk

QUESTION 1

Let the disk be  

$$
\Omega=\{(x,y)\in \mathbb{R}^2 : x^2+y^2\le 10^2\}.
$$

Let $A$ be the event that the chosen point lies within 1 unit of the boundary.

$$
A=\{(x,y)\in \Omega : 9 \le \sqrt{x^2+y^2} \le 10\}
$$

---

## Uniform Distribution with Respect to Area

Total area of the disk:

$$
100\pi
$$

Area of annulus:

$$
100\pi - 81\pi = 19\pi
$$

Probability:

$$
P(A) = \frac{19\pi}{100\pi} = \frac{19}{100}
$$

---

## Uniform Distribution of the Radius

Total radius length:

$$
10
$$

Favorable length:

$$
1
$$

Probability:

$$
P(A) = \frac{1}{10}
$$

---

## Why Answers Differ

- Area-uniform → depends on **area**
- Radius-uniform → depends on **distance**

$$
P(A)=\frac{19}{100} \quad \text{vs} \quad P(A)=\frac{1}{10}
$$

---

## Conclusion

- Real-world (like dart throwing) → **area model**
- Artificial method → **radius model**



\# Bertrand's Paradox: Random Chords
QUESTION 2

## Preliminary Geometric Fact

Let the circle be:

$$
x^2 + y^2 = 1
$$

The side length of the inscribed equilateral triangle is:

$$
\sqrt{3}
$$

A chord is longer than $\sqrt{3}$ if and only if its perpendicular distance from the center is less than $1/2$.

---

## Proof

If a chord is at distance $r \in [0,1]$ from the center, then half its length is:

$$
L(r) = 2\sqrt{1 - r^2}
$$

We want:

$$
L(r) > \sqrt{3}
$$

This gives:

$$
r < \frac{1}{2}
$$

So the event is:

$$
A = \{\text{distance of chord from center} < 1/2\}
$$

---

## Model 1: Uniform Endpoint / Angle Model

### Probability Space

$$
\Omega_1 = [0,\pi], \quad P_1(E) = \frac{\lambda(E)}{\pi}
$$

Let $\theta \in [0,\pi]$ be the central angle.

### Condition

Chord length:

$$
L(\theta) = 2\sin(\theta/2)
$$

We require:

$$
2\sin(\theta/2) > \sqrt{3}
\Rightarrow \theta > \frac{2\pi}{3}
$$

So:

$$
A_1 = (2\pi/3, \pi]
$$

### Probability

$$
P_1 = \frac{\pi - 2\pi/3}{\pi} = \frac{1}{3}
$$

---

## Model 2: Uniform Distance Model

$$
\Omega_2 = [0,1], \quad P_2(E) = \lambda(E)
$$

Event:

$$
A_2 = [0,1/2)
$$

Probability:

$$
P_2 = \frac{1}{2}
$$

---

## Model 3: Uniform Midpoint Model

$$
\Omega_3 = \{(x,y): x^2 + y^2 \le 1\}
$$

$$
P_3(E) = \frac{\lambda_2(E)}{\pi}
$$

Event:

$$
x^2 + y^2 < \frac{1}{4}
$$

Probability:

$$
P_3 = \frac{\pi(1/2)^2}{\pi} = \frac{1}{4}
$$

---

## Conclusion

| Model | Method | Probability |
|------|--------|-------------|
| 1 | Angle | 1/3 |
| 2 | Distance | 1/2 |
| 3 | Midpoint | 1/4 |

---

### Final Insight

Different definitions of “random chord” give different results. This is Bertrand’s paradox.

\## Conditional Probability (Two Children Problem)
QUESTION 8

The sample space is:

$$
\Omega = \{BB, BG, GB, GG\}
$$

where the first letter denotes the older child and the second the younger.

Each outcome has probability:

$$
\frac{1}{4}
$$

---

### Step 1: Apply the condition

The condition “at least one is a girl” excludes only $BB$.

So the reduced sample space becomes:

$$
\{GG, BG, GB\}
$$

These are equally likely.

---

### Step 2: Count favorable outcomes

Only one outcome corresponds to both girls:

$$
GG
$$

So,

$$
P(\text{both girls} \mid \text{at least one girl}) = \frac{1}{3}
$$

---

### Final Answer

$$
\boxed{\frac{1}{3}}
$$

---

## Why this is interesting

Many people incorrectly guess $\frac{1}{2}$.

That answer would correspond to a different type of information, for example:

> “The older child is a girl.”

Then the sample space becomes:

$$
\{GG, GB\}
$$

So:

$$
P(\text{both girls} \mid \text{older child is a girl}) = \frac{1}{2}
$$

---

### Key Insight

Probability depends not only on the event, but also on **how the information is given**.

\## Probability Space and Random Variables
QUESTION 9

### 1. Probability Space

The probability space is:

$$
\Omega = \{G_1, G_2, D_1, D_2\},
$$

$$
\mathfrak{F} = \mathfrak{P}(\Omega),
$$

with probability measure defined as:

$$
P(\{G_1\}) = 0.50,\quad P(\{G_2\}) = 0.20,\quad P(\{D_1\}) = 0.10,\quad P(\{D_2\}) = 0.20.
$$

Since:

$$
0.50 + 0.20 + 0.10 + 0.20 = 1,
$$

this is a valid probability measure.

---

## 2. Random Variable X

The random variable $X$ records whether an item is good or defective:

$$
X(G_1) = 0,\quad X(G_2) = 0,\quad X(D_1) = 1,\quad X(D_2) = 1.
$$

Inverse images:

$$
X^{-1}(\{0\}) = \{G_1, G_2\},
$$

$$
X^{-1}(\{1\}) = \{D_1, D_2\}.
$$

Thus,

$$
\sigma(X) = \{\varnothing, \Omega, \{G_1,G_2\}, \{D_1,D_2\}\}.
$$

---

## 3. Random Variable Y

The random variable $Y$ records both quality and machine:

$$
Y(G_1) = 1,\quad Y(G_2) = 2,\quad Y(D_1) = 3,\quad Y(D_2) = 4.
$$

Since each outcome is uniquely identified:

$$
\sigma(Y) = \mathfrak{P}(\Omega).
$$

Explicitly:

$$
\sigma(Y) = \{\varnothing, \{G_1\}, \{G_2\}, \{D_1\}, \{D_2\}, \{G_1,G_2\}, \{G_1,D_1\}, \ldots, \Omega\}.
$$

---

## 4. Relationship Between σ-algebras

Since:

$$
\sigma(X) \subseteq \sigma(Y),
$$

the variable $Y$ contains more information than $X$.

---

## 5. Distribution of X

$$
P(X=0) = P(\{G_1,G_2\}) = 0.50 + 0.20 = 0.70,
$$

$$
P(X=1) = P(\{D_1,D_2\}) = 0.10 + 0.20 = 0.30.
$$

So:

$$
P(X=0)=0.70,\quad P(X=1)=0.30.
$$

---

## 6. Distribution of Y

$$
P(Y=1)=0.50,\quad P(Y=2)=0.20,\quad P(Y=3)=0.10,\quad P(Y=4)=0.20.
$$

---

## 7. Joint Distribution

$$
P(X=0,Y=1)=0.50,\quad P(X=0,Y=2)=0.20,
$$

$$
P(X=1,Y=3)=0.10,\quad P(X=1,Y=4)=0.20.
$$

All other joint probabilities are zero.

---

## 8. Conditional Probabilities

$$
P(X=1 \mid Y=3) = 1,\quad P(X=1 \mid Y=4) = 1.
$$

Now:

$$
P(Y=3 \mid X=1) = \frac{0.10}{0.30} = \frac{1}{3},
$$

$$
P(Y=4 \mid X=1) = \frac{0.20}{0.30} = \frac{2}{3}.
$$

---

## 9. Interpretation

Knowing $Y$ gives more precise information than knowing $X$.

If $Y=3$, then the item is certainly defective:

$$
P(X=1 \mid Y=3)=1.
$$

But if $X=1$, uncertainty remains about the machine.

---

## 10. Conclusion

The $\sigma$-algebra generated by $X$ only distinguishes:

- good vs defective

while $\sigma(Y)$ identifies every outcome.

Therefore:

$$
\sigma(X) \subseteq \sigma(Y).
$$

So $Y$ carries strictly more information than $X$.

\## Shannon Entropy Calculations
QUESTION 9

Using natural logarithms, the Shannon entropy is:

$$
H(Z) = -\sum_z P(z)\ln P(z)
$$

---

### Entropy of X

$$
H(X) = -(0.70\ln 0.70 + 0.30\ln 0.30)
$$

$$
H(X) \approx 0.61086
$$

---

### Entropy of Y

$$
H(Y) = -(0.50\ln 0.50 + 0.20\ln 0.20 + 0.10\ln 0.10 + 0.20\ln 0.20)
$$

$$
H(Y) \approx 1.22061
$$

---

### Joint Entropy

$$
H(X,Y) = -\sum_{x,y} P(x,y)\ln P(x,y)
$$

Only non-zero terms contribute:

$$
H(X,Y) = -(0.50\ln 0.50 + 0.20\ln 0.20 + 0.10\ln 0.10 + 0.20\ln 0.20)
$$

$$
H(X,Y) \approx 1.22061
$$

---

### Conditional Entropies

Since each value of $Y$ determines $X$ exactly:

$$
H(X \mid Y) = 0
$$

Also:

$$
H(Y \mid X) = H(X,Y) - H(X)
$$

$$
H(Y \mid X) \approx 1.22061 - 0.61086 = 0.60975
$$

---

### Final Summary

$$
H(X) \approx 0.61086
$$

$$
H(Y) \approx 1.22061
$$

$$
H(X,Y) \approx 1.22061
$$

$$
H(X \mid Y) = 0
$$

$$
H(Y \mid X) \approx 0.60975
$$

---

## Interpretation

- $H(X)$ measures uncertainty about good vs defective items  
- $H(Y)$ measures uncertainty about full classification  
- $H(X \mid Y)=0$ because $Y$ fully determines $X$  
- $H(Y \mid X)>0$ because $X$ does not determine machine  

### Key Insight

This shows mathematically that $Y$ contains more information than $X$.

\# Probability Space with Temperature and Pressure QUESTION 10

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, multivariate_normal

# Parameters
mu_T, sigma_T = 0.0, 1.0
mu_P, sigma_P = 0.0, 2.0   # Var(P)=4

# Grids
t = np.linspace(-4, 4, 400)
p = np.linspace(-8, 8, 400)
T, P = np.meshgrid(t, p)

# Joint density of Y = (T, P)
joint_pdf = (
    norm.pdf(T, loc=mu_T, scale=sigma_T) *
    norm.pdf(P, loc=mu_P, scale=sigma_P)
)

# Marginal density of X = T
fx = norm.pdf(t, loc=mu_T, scale=sigma_T)

# Conditional density P | X=t0
t0 = 1.0
cond_p_given_x = norm.pdf(p, loc=mu_P, scale=sigma_P)

# Conditional visualization X | Y=(t1,p1)
t1, p1 = 1.0, 2.0
eps = 0.08
cond_x_given_y_vis = norm.pdf(t, loc=t1, scale=eps)

# -------------------------
# NEW STYLE LAYOUT
# -------------------------
fig = plt.figure(figsize=(13, 11))
fig.patch.set_facecolor("#f7f9fc")

gs = fig.add_gridspec(
    3, 2,
    height_ratios=[1.15, 1, 1],
    hspace=0.5,
    wspace=0.3
)

# -------------------------
# (1) Joint density (TOP)
# -------------------------
ax1 = fig.add_subplot(gs[0, :])
ax1.set_facecolor("#f0f3f7")

cs = ax1.contourf(
    T, P, joint_pdf,
    levels=30,
    cmap="magma"
)

ax1.contour(
    T, P, joint_pdf,
    levels=8,
    colors="#ffffff",
    linewidths=0.6,
    alpha=0.5
)

ax1.set_title(r"(1) Joint density of $Y=(T,P)$", fontsize=13, color="#2c3e50")
ax1.set_xlabel(r"$t$")
ax1.set_ylabel(r"$p$")

cbar = fig.colorbar(cs, ax=ax1, shrink=0.92)
cbar.outline.set_visible(False)
cbar.set_label("Density")

ax1.grid(alpha=0.15, linestyle="--")

# -------------------------
# (3) Marginal of X
# -------------------------
ax3 = fig.add_subplot(gs[1, 0])
ax3.set_facecolor("#ffffff")

ax3.plot(t, fx, color="#1f77b4", lw=2.2, linestyle="-")
ax3.fill_between(t, fx, color="#aec7e8", alpha=0.45)

ax3.set_title(r"(3) Marginal density of $X=T$", fontsize=12)
ax3.set_xlabel(r"$x$")
ax3.set_ylabel(r"$f_X(x)$")
ax3.grid(alpha=0.2, linestyle=":")

# -------------------------
# (4) Marginal of P
# -------------------------
ax4 = fig.add_subplot(gs[1, 1])
ax4.set_facecolor("#ffffff")

fp = norm.pdf(p, loc=mu_P, scale=sigma_P)

ax4.plot(p, fp, color="#2ca02c", lw=2.2, linestyle="--")
ax4.fill_between(p, fp, color="#98df8a", alpha=0.4)

ax4.set_title(r"(4) Marginal density of $P$", fontsize=12)
ax4.set_xlabel(r"$p$")
ax4.set_ylabel(r"$f_P(p)$")
ax4.grid(alpha=0.2, linestyle=":")

# -------------------------
# (5) Conditional P | X
# -------------------------
ax5 = fig.add_subplot(gs[2, 0])
ax5.set_facecolor("#ffffff")

ax5.plot(
    p, cond_p_given_x,
    color="#d62728",
    lw=2.3,
    linestyle="-.",
    label=rf"$P \mid X={t0}$"
)

ax5.fill_between(p, cond_p_given_x, color="#ff9896", alpha=0.35)

ax5.set_title(r"(5) Conditional density $P \mid X=t_0$", fontsize=12)
ax5.set_xlabel(r"$p$")
ax5.set_ylabel("Density")
ax5.legend(frameon=False)
ax5.grid(alpha=0.2, linestyle=":")

# -------------------------
# (6) X | Y visualization
# -------------------------
ax6 = fig.add_subplot(gs[2, 1])
ax6.set_facecolor("#ffffff")

ax6.plot(
    t, cond_x_given_y_vis,
    color="#9467bd",
    lw=2.4,
    linestyle="-"
)

ax6.axvline(
    t1,
    color="#000000",
    linestyle="--",
    lw=1.2
)

ax6.set_title(r"(6) $X \mid Y=(t_1,p_1)$ (degenerate at $t_1$)", fontsize=12)
ax6.set_xlabel(r"$x$")
ax6.set_ylabel("Approx. density")

ax6.grid(alpha=0.2, linestyle=":")

plt.show()
```
